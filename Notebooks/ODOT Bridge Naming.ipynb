## Animal

Determines the route carrying the bridge, higher order animals are  
associated with more important routes,

1. Interstate route - Chordata
2. U.S. route - Chordata
3. State route - Chordata
4. County route - Arthropoda
5. City street - Mollusca
6. Federal lands road - Cnidaria
7. State lands road - Echinodermata
8. Other - Porifera

It's understood that people are mostly unfamiliar with the animals outside  
of Chordata. Since it's an ODOT system, and users will mostly be working  
with State Routes or above, the other categories were given minor animals.

The chordata class is further broken up into 

Interstate Routes ~ 71 - Mammals (Only Ohio Routes shown for all)  
US Routes ~ 300 - Birds  
State Routes ~ 3000 - Reptiles, amphibians and fish

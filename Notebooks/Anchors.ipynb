# Concrete Anchors (Design Guides)

Based on ACI Chapter 17

## ACI 318-19 Chapter 17 - Anchoring to Concrete

### 17.1 - Scope

**17.1.2** Provisions apply to the following anchor types (a) through (g):

(a) Headed studs and headed bolts having a geometry that has been
demonstrated to result in a pullout strength in uncracked concrete equal
to or exceeeding $1.4N_P$ where $N_P$ is given in Eq. (17.6.3.2.2a). <!-- #//TODO - LINK -->

(b) Hooked bolts having a geometry that has been demonstrated to result in
a pullout strength without the benefit of friction in uncracked concrete
equal to or exceeding $1.4N_p$, where $N_P$ is given in Eq. (17.6.3.2.2b) <!-- #//TODO - LINK -->

(c) Post-installed expansion (torque controlled and
displacement-controlled) anchors that meet the assessment criteria of ACI
355.2. <!-- #//TODO - LINK -->

(d) Post-installed undercut anchors that meet the assessment criteria of
ACI 355.2. <!-- #//TODO - LINK -->

(e) Post-installed adhesive anchors that meet the new assessment criteria
of ACI 355.4. <!-- #//TODO - LINK -->

(f) Post-installed screw anchors that meet the assessment criteria of ACI
355.2. <!-- #//TODO - LINK -->

(g) Attachments with shear lugs.

**17.1.3** The removal and resetting of post-installed mechanical anchors
is prohibited.

**17.1.4** This chapter does not apply for load applications that are
predominantly high-cycle fatigue or due to impact.

**17.1.5** This chapter does not apply to specialty inserts, through-bolts,
multiple anchors connecto to a single steel plate at the embedded end of
the anchors, grouted anchors, or power driven anchors such as powder or
pneumatic actuated fasteners.

**17.1.6** Reinforcement used as part of an embedment shall have
development length established in accordance with other parts of this Code.
If reinforcement is used as anchorage, concrete breakout failure shall be
considered. Alternatively, anchor reinforcement in accordance with 17.5.2.1 <!-- #//TODO - LINK -->
shall be provided.

### 17.2 - General

**17.2.1** Anchors and anchor groups shall be designed for critical effects
of factored loads calculated by elastic analysis. If nominal strength is
controlled by ductile steel elements, plastic analysis is permitted
provided that deformation compatibility is taken into account.

**17.2.1.1** Anchor group effects shall be considered if two or more
anchors loaded by a common structural element are spaced closer than the
spacing required for unreduced breakout strength. If adjacent anchors are
not loaded by a common structural element, group effects shall consider
simultaneous maximum loading of adjacent anchors.

**17.2.2** Adhesive anchors shall be installed in concrete having a minimum
age of 21 days at time of anchor installation.

**17.2.3** Adhesive anchors installed horizontally or upwardly inclined
shall be qualified in accordance with ACI 355.4 requirements for
sensitivity to installation direction.

**17.2.4** *Lightweight concrete modification factor*, $ \lambda_a $

**17.2.4.1** Modification factor $ \lambda_a $ for lightweight concrete shall be in
accordance with Table 17.2.4.1. It shall be permitted to use an alternate
value of $ \lambda_a $ if tests are performed and evaluated in accordance with ACI
355.2 or ACI 355.4.

**Table 17.2.4.1 - Modification factor $\lambda_a$ for lightweight concrete**

|                 Case                                   | $\lambda_a^{[1]}$ |
|:------------------------------------------------------:|:-----------------:|
| Cast-in and undercut anchor concrete failure           | 1.0 $ \lambda $   |
| Expansion, screw, and adhesive anchor concrete failure | 0.8 $ \lambda $   |
| Adhesive anchor bond failure per Eq. (17.6.5.2.1)      | 0.6 $ \lambda $   |

<center><sup>[1]</sup>$\lambda$ shall be in accordance with 19.2.4</center>

**17.2.5** Anchors shall be installed and inspected in accordance with the
requirements of 26.7 and 26.13.

### 17.3 - Design Limits



**17.3.1** The value of $f_c'$ used for calculation purposes in this chapter
shall not exceed 10,000 psi for cast-in anchors and 8000 psi for
post-installed anchors. Post-installed anchors shall not be used in
concrete with a strength greater than 8000 psi without testing to verify
acceptable performance.

**17.3.2** For anchors with diameters $d_a \leq 4\ \text{in.}$, concrete breakout
strength requirements shall be considered satisfied by the design
procedures of 17.6.2 and 17.7.2.

**17.3.3** For adhesive anchors with embedment depths $4d_a \leq h_{ef} \leq 20d_a $,
bond strength requirements shall be considered satisfied by the design
procedure of 17.6.5.

**17.3.4** For screw anchors with embedment depths $5d_a \leq h_{ef} \leq 10d_a $
and $h_{ef} \geq 1.5\ in. $, concrete breakout strength requirements shall
be considered satisfied by the design procedures of 17.6.2 and 17.7.2.

**17.3.5** Anchors shall satisfy the edge distances, spacings, and
thicknesses in 17.9 unless supplementary reinforcement is provided to
control splitting failure.

### 17.4 - Required Strength

**17.4.1** Required strength shall be calculated in accordance with the
factored load combinations in Chapter 5.

**17.4.2** For anchors in structures assigned to SDC C, D, E, and F, the
additional requirements of 17.10 shall apply.

### 17.5 - Design Strength

**17.5.1** For each applicable factored load combination, design strength
of individual anchors and anchor groups shall satisfy $\phi S_n \geq U $.
Interaction between load effects shall be considered in accordance with
17.8.1.

**17.5.1.1** Strength reduction factor, $\phi$, shall be determined in
accordance with 17.5.3.

**17.5.1.2** Nominal strength for an anchor or anchor groups shall be based
on design models that result in predictions of strength in substantial
agreement with results of comprehensive tests. The materials used in the
tests shall be compatible with the materiasl used in the structure. The
nominal strength shall be based on the 5 percent fractile of the basic
individual anchor strength. For nominal strengths related to concrete
strength, modifications for size effects, number of anchors, effects of
close spacing of anchors, proximity ot edges, depth of the concrete member,
eccentric loadings of anchor groups, and influence of cracking shall be
taken into account. Limits on edge distance and anchor spacing in the
design models shall be consistent with the tests that verified the model.
Strength of anchors shall be based on design models that satisfy 17.5.1.2
for the following:

(a) Steel strength of anchor in tension
(b) Concrete breakout strength of anchor in tension
(c) Pullout strength of a single cast-in anchor and single post-installed
expansion, screw, and undercut anchor in tension
(d) Concrete side-face blowout strength of headed anchor in tension
(e) Bond strength of adhesive anchor in tension
(f) Steel strength of anchor in shear
(g) Concrete breakout strength of anchor in shear
(h) Concrete pryout strength of anchor in shear

(See ACI 318 Fig. R17.5.1.2 for anchor failure modes)

**17.5.1.3** Strength of anchors shall be permitted to be determined in
accordance with 17.6 for 17.5.1.2(a) through (e), and 17.7 for 17.5.1.2(f)
through (h). For adhesive anchors that resist sustained tension, the
requirements of 17.5.2.2 shall apply.

**17.5.1.3.1** Anchor group effects shall be considered wherever two or
more anchors have spacing less than the critical spacing in Table
17.5.1.3.1, where only those anchors susceptible to the particular failure
mode under investigation shall be included in the group.

**Table 17.5.1.3.1 - Critical Spacing**

| Failure mode under investigation | Critical Spacing |
|:--------------------------------:|:----------------:|
| Concrete breakout in tension     |    $3h_{ef}$     |
| Bond strength in tension         |    $2c_{Na}$     |
| Concrete breakout in shear       |    $3c_{a1}$     |

**17.5.1.4** Strength of anchors shall be permitted to be based on test
evaluation using the 5 percent fractile of applicable test results for
17.5.1.2 (a) through (h).

**17.5.2** Fore each applicable factored load combination, design stregnth
of anchors shall satisfy the criteria in Table 17.5.2.

**17.5.2.4** Anchors shall satisfy the edge distances, spacings and

**17.5.2.5** Anchors in structures assigned to Seismic Design Category C,

**17.5.2.6** Attachments with shear lugs used to transfer structural loads
shall satisfy the requirements of 17.11.

### 17.6 - Tensile Strength

**17.6.1** *Steel strength of anchors in tension*, $N_{sa}$

**17.6.1.1** Nominal steel strength of anchors in tension as governed by
the steel, $N_{sa}$, shall be evaluated based on the properties of the
anchor material and the physical dimensions of the anchors.

**17.6.1.2** Nominal steel strength of an anchor in tension, $N_{sa}$,
shall be calculated by:

$$ N_{sa} = A_{se,N}f_{uta} \tag{17.6.1.2} $$

where $A_{se,N}$ is the effective cross-sectional area of an anchor in
tension, in.<sup>2</sup>, and $f_{uta}$ used for calculations shall not exceed either
$1.9f_{ya}$ or 125,000 psi.

**17.6.2** *Concrete breakout strength of anchors in tension*, $N_{cb}$

**17.6.2.1** Nominal concrete breakout strength in tension, $N_{cb}$ of a
single anchro or $N_{cbg}$ of an anchor grou satisfying 17.5.1.3.1, shall
be calculated by (a) or (b) respectively:

(a) For a single anchor

$$ N_{cb} = \frac{A_{Nc}}{A_{Nco}} \psi_{ed,N}\ \psi_{c,N}\ \psi_{cp,N} N_b \tag{17.6.2.1a} $$

(b) For an anchor group

$$ N_{cbg} = \frac{A_{Nc}}{A_{Nco}} \psi_{ec,N}\ \psi_{ed,N}\ \psi_{c,N}\ \psi_{cp,N} N_b \tag{17.6.2.1b} $$

where $\psi_{ec,N}\ \psi_{ed,N}\ \psi_{c,N}$ and $\psi_{cp,N}$ are given in
17.6.2.3, 17.6.2.4, 17.6.2.5, and 17.6.2.6, respectively.

**17.6.2.1.1** $A_{Nc}$ is the projected concrete failure area of a single
anchor or of an anchor group that is approximated as the base of the
rectilinear geometrical shape that results from projecting the failure
surface outward $1.5h_{ef}$ from the centerlines of the anchor, or in the
case of an anchor group, from a line through a row of adjacent anchors.
$A_{Nc}$ shall not exceed $nA_{Nco}$ where $n$ is the number of anchors in
the group that resist tension.

**17.6.2.1.2** If anchors are located less than $1.5h_{ef}$ from three or
more edges, the value of $h_{ef}$ used to calculate A_{Nc} in accordance
with 17.6.2.1.1, as well as for the equations in 17.6.2.1 throuhg 17.6.2.4,
shall be the greater of (a) and (b):

&nbsp;&nbsp;&nbsp;&nbsp;(a) $c_{a,max}/1.5$
&nbsp;&nbsp;&nbsp;&nbsp;(b) $s/3$ where $s$ is the maximum spacing between anchors within the group.

**17.6.2.1.3** If an additional plate or washer is added at the head of the
anchor, it shall be permitted to calculate the projected area of the
failure surface by projecting the failure surface outward $1.5h_{ef}$ from
the effective perimeter of the plate or washer. The effective perimeter
shall not exceed the value at a section projected outward more than the
thickness of the washer or plate from the outer edge of the head of the
anchor.

**17.6.2.1.4** $A_{Nco}$ is the projected concrete failure area of a single
anchor with an edge distance of at least $1.5h_{ef}$ and shall be calculated
by Eq. (17.6.2.1.4).

$$ A_{Nco} = 9h_{ef}^2 \tag{17.6.2.1.4} $$

**17.6.2.2** *Basic single anchor breakout strength*, $N_b$

**17.6.2.2.1** Basic concrete breakout strength of a single anchor in
tension in cracked concrete, $N_b$, shall be calculated by Eq.
(17.6.2.2.1), except as permitted in 17.6.2.2.3

$$ N_b = k_c \lambda_a \sqrt{f_c'}h_{ef}^{1.5} \tag{17.6.2.2.1} $$

where $k_c = 24$ for cast-in anchors and **17** for post-installed anchors.

**17.6.2.2.2** $k_c$ for post-installed anchors shall be permitted to be
increased based on ACI 355.2 or ACI 355.4 product-specific tests, but shall
not exceed 24.

**17.6.2.2.3** For single cast-in headed studs and headed bolts with
$ 11\ in. \leq h_{ef} \leq 25\ in. $, $N_b$ shall be calculated by:

$$ N_b = 16 \lambda_a \sqrt{f_c'} h_{ef}^{5/3} \tag{17.6.2.2.3} $$

**17.6.2.3** *Breakout eccentricity factor*, $ \psi_{ec,N} $

**17.6.2.3.1** Modification factor for anchor groups loaded eccentrically
in tension, $\psi_{ec,N}$, shall be calculated by Eq. (17.6.2.3.1).

$$ \psi_{ec,N} = \frac{1}{1+\frac{e_N'}{1.5h_{ef}}} \leq\ 1.0 \tag{17.6.2.3.1} $$

**17.6.2.3.2** If the loading on an anchor group is such that only some of
the anchors in the group are in tension, only those anchors that are in
tension shall be considered for determining eccentricity $e_n'$ in Eq.
(17.6.2.3.1) and for the calculation of $N_{cbg}$ according to Eq. (17.6.2.1b).

**17.6.2.4** *Breakout edge effect factor*, $\psi_{ed,N}$

**17.6.2.4.1** Modification factor for edge effects for single anchors or
anchor groups loaded in tension, $\psi_{ed,N}$, shall be determined by (a)
or (b).

$$(a)\text{If}\ c_{a, min}\ \geq 1.5h_{ef},\text{then}\ \psi_{ed,N}\ =\ 1.0 \tag{17.6.2.4.1a} $$

$$(b)\text{If}\ c_{a, min}\ < 1.5h_{ef},\text{then}\ \psi_{ed,N}\ =\ 0.7 + 0.3\frac{c_{a,min}}{1.5h_{ef}} \tag{17.6.2.4.1b} $$

**17.6.2.5** *Breakout cracking factor*, $\psi_{c,N}$

**17.6.2.5.1** Modification factor for the influence of cracking in anchor
regions at service load levels, $\psi_{c,N}$ shall be determined by (a) or
(b):

&nbsp;&nbsp;(a) For anchors located in a region of a concrete member where
&nbsp;&nbsp;analysis indicates no cracking at service load levels, $\psi_{c,N}$ shall
&nbsp;&nbsp;be permitted to be:
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$\psi_{c,N}=1.25$ for cast-in anchors
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$\psi_{c,N}=1.4$ for post-installed anchors, if the value of $k_c$ used
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;in Eq. (17.6.2.2.1) is 17. If the value of $k_c$ used in Eq. (17.6.2.2.1)
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;is taken from the ACI 355.2 or CI 355.4 product evaluation report
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;for post-installed anchors:
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(i) $\psi_{c,N}$ shall be based on the ACI 355.2 or ACI 355.4 product
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;evaluation report for anchors qualified for use in both cracked and
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;uncracked concrete.
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(ii) $\psi_{c,N}$ shall be taken as 1.0 for anchors qualified for
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;use in uncracked concrete.
&nbsp;&nbsp;(b) For anchors located in a region of a concrete member where
&nbsp;&nbsp;analysis indicates cracking at service load levels, $\psi_{c,N}$ shall be
&nbsp;&nbsp;taken as 1.0 for both cast-in anchors and post-installed anchors, and
&nbsp;&nbsp;17.6.2.6 shall be satisfied.

**17.6.2.5.2** Post-installed anchors shall be qualified for use in cracked
concrete in accordance with ACI 355.2 or ACI 355.4. Cracking in the
concrete shall be controlled by flexural reinforcement distributed in
accordance with 24.3.2, or equivalent crack control shall be provided by
confining reinforcement.

**17.6.2.6** *Breakout splitting factor*, $\psi_{cp,N}$

**17.6.2.6.1** Modification factor for post-installed anchors designed for
uncracked concrete in accordance with 17.6.2.5 without supplementary
reinforcement to control splitting, $\psi_{cp,N}$, shall be determined by
(a) or (b) using the critical distance $c_{ac}$ as defined in 17.9.5.

$$(a)\text{If}\ c_{a, min}\ \geq c_{ac}, \text{then}\ \psi_{cp,N}\ =\ 1.0 \tag{17.6.2.6.1a} $$

$$(b)\text{If}\ c_{a, min}\ < c_{ac}, \text{then}\ \psi_{cp,N}\ =\ \frac{c_{a,min}}{c_{ac}} \geq \frac{1.5h_{ef}}{c_{ac}} \tag{17.6.2.4.1b} $$

**17.6.2.6.2** For all other cases, including cast-in anchors, $\psi_{cp,N}$
shall be taken as 1.0.

**17.6.3** Pullout strength of a single cast-in anchor or a single
post-installed expansion, screw or undercut anchor in tension, $N_{pn}$

**17.6.3.1** Nominal pullout strength of a single cast-in anchor or a
single-post-installed expansion, screw, or undercut anchor in tension,
$N_{pn}$, shall be calculated by:

$$ N_{pn} = \psi_{c,P} N_P \tag{17.6.3.1} $$

where $\psi_{c,P}$ is given in 17.6.3.3.

**17.6.3.2** *Basic single anchor pullout strength*, $N_p$

**17.6.3.2.1** For post-installed expansion, screw, and undercut anchors,
the values of $N_p$ shall be based on the 5 percent fractile of results of
tests performed and evaluated according to ACI 355.2. It is not permissible
to calculate the pullout strength in tension for such anchors.

**17.6.3.2.2** For single anchors, it shall be permitted to evaluate the
pullout strength in tension, $N_p$, for use in Eq. (17.6.3.1) in accordance
with (a) or (b). Alternatively, it shall be permitted to use values of $N_p$
based on the 5 percent fractile of tests performed and evaluated in the
same manner as the ACI 355.2 procedures but without the benefit of
friction.

(a) For cast-in headed studs and headed bolts, $N_p$ shall be calculated by:

$$ N_p = 8A_{brg}f_c' \tag{17.6.3.2.2a} $$

(b) For J- or L-bolts, $N_p$ shall be calculated by:

$$ N_p = 0.9f_c' e_h d_a \tag{17.6.3.2.2b} $$

where $3d_a \leq e_h \leq 4.5d_a$.

**17.6.3.3** *Pullout cracking factor* $\psi_{c,P}$

**17.6.3.3.1** Modification factor to account for the influence of cracking
in anchor regions at service load levels, $\psi_{c,P}$, shall be determined
by (a) or (b):

(a) For anchors located in a region of a concrete member where analysis
indicates no cracking at service load levels, $\psi_{c,P}$ shall be taken
as 1.4.

(b) For anchors located ina region of a concrete member where analysis
indicates cracking at service load levels, $\psi_{c,P}$, shall be taken as
1.0.

**17.6.4** *Concrete side-face blowout strength of headed anchors in tension,* $N_{sb}$

**17.6.4.1** For a single headed anchor with deep embedment close to an
edge $(h_{ef} > 2.5 c_{a1})$, the nominal side-flace blowout strength,
$N_{sb}$, shall be calculated by:

$$ N_{sb} = 160 c_{a1} \sqrt{A_{brg}} \lambda_a \sqrt{f_c'} \tag{17.6.4.1} $$

**17.6.4.1.1** If $c_{a2}$ for the single headed anchor is less than $3c_{a1}$,
the value of $N_{sb}$, shall be multiplied by the factor $(1 + c_{a2} / c_{a1})/4$,
where $1.0 \leq c_{a2}/c_{a1} \leq 3.0.$


**17.6.4.2** For multiple headed anchors with deep embedment close to an
edge ($h_{ef}>2.5c_{a1}$) and anchor spacing less than $6c_{a1}$, the
nominal strength of those anchors susceptible to a side-face blowout
failure, $N_{sbg}$, shall be calculated by:

$$ N_{sbg} = (1 + \frac{s}{6c_{a1}}) N_{sb} \tag{17.6.4.2} $$

where $s$ is the distance between the outer anchors along the edge, and
$N_{sb}$ is obtained from Eq. (17.6.4.1) without modification for a
perpendicular edge distance.

**17.6.5** *Bond strength of adhesive anchors in tension*, $N_a$ *or* $N_{ag}$

**17.6.5.1** Nominal bond strength in tension, $N_a$ of a single adhesive
anchor or $N_{ag}$ of an adhesive anchor group satisfying 17.5.1.3.1, shall
be calculated by (a) or (b), respectively.

&nbsp;&nbsp;(a) For a single adhesive anchor:

$$ N_a = \frac{A_{Na}}{A_{Nao}}\ \psi_{ed,Na}\ \psi_{cp,Na}\ N_{ba} \tag{17.6.5.1a} $$

&nbsp;&nbsp;(b) For an adhesive anchor group:

$$ N_{ag} = \frac{A_{Na}}{A_{Nao}}\ \psi_{ec,Na}\ \psi_{ed,Na}\ \psi_{cp,Na}\ N_{ba} \tag{17.6.5.1b} $$

where $\psi_{ec,Na}\ \psi_{ed,Na}$ and $\psi_{cp,Na}$ are given in 17.6.5.3, 17.6.5.4, and
17.6.5.5, respectively.

**17.6.5.1.1** $A_{Na}$ is the projected influence area of a single
adhesive anchor or an adhesive anchor group that is approximated as a
rectilinear area that projects outward a distance $c_{Na}$ from the
centerline of the adhesive anchor, or in the case of an adhesive anchor
group, from a line through a row of adjacent adhesive anchors. $A_{Na}$
shall not exceed $nA_{Nao}$, where *n* is the number of adhesive anchors in
the group that resist tension.

**17.6.5.1.2** $A_{Nao}$ is the projected influence area of a single
adhesive anchor with an edge distance of at least $c_{Na}$:

$$ A_{Nao} = (2c_{Na})^2 \tag{17.6.5.1.2a} $$

where:

$$ c_{Na} = 10 d_a \sqrt{\frac{\tau_{uncr}}{1100}} \tag{17.6.5.1.2b} $$

**17.6.5.2** *Basic single anchor bond strength*, $N_{ba}$

**17.6.5.2.1** Basic bond strength of a single adhesive anchor in tension
in cracked concrete, $N_{ba}$, shall be calculated by Eq. (17.6.5.2.1)

$$ N_{ba} = \lambda_a \tau_{cr} \pi d_a h_{ef} \tag{17.6.5.2.1} $$

**17.6.5.2.2** Characteristic bond stress, $\tau_{cr}$, shall be taken as
the 5 percent fractile of results of tests performed and evaluated in
accordance with ACI 355.4.

**17.6.5.2.3** If analysis inidcates cracking at service load levels,
adhesive anchors shall be qualified for use in cracked concrete in
accordance with ACI 355.4.

**17.6.5.2.4** For adhesive anchors located in a region of a concrete member
where analysis indicates no cracking at service load levels, $\tau_{uncr}$ shall
be permitted to be used in place of $\tau_{cr}$ in Eq. (17.6.5.2.1) and shall
be taken as the 5 percent fractile of results of tests performed and
evaluated according to ACI 355.4.

**17.6.5.2.4** For adhesive anchors located in a region of a concrete member
where analysis indicates no cracking at service load levels, $\tau_{uncr}$
shall be permitted to be used in place of $\tau_{cr}$ in Eq. (17.6.5.2.1)
and shall be taken as the 5 percent fractile of results of tests performed
and evaluated according to ACI 355.4.

**17.6.5.2.5** It shall be permitted to use the minimum characteristic bond
stress values in Table 17.6.5.2.5, provided (a) through (e) are satisfied:

&nbsp;&nbsp;(a) Anchors shall meet the requirements of ACI 355.4
&nbsp;&nbsp;(b) Anchors shall be installed in holes drilled with a rotary impact drill
&nbsp;&nbsp;or rock drill
&nbsp;&nbsp;(c) Concrete compressive strength at time of anchor installation shall be
&nbsp;&nbsp;at least 2500 psi
&nbsp;&nbsp;(d) Concrete age at time of anchor installation shall be at least 21 days
&nbsp;&nbsp;(e) Concrete temperature at time of anchor installation shall be at least
&nbsp;&nbsp;50°F

|:------------------------------------:|:-----------------------------------------------------------:|:-------------------------------------------:|:-----------:|:------------------:|
| Outdoor                              |  Dry to fully saturated                                     |  175                                        | 200         |    650             |

**17.6.5.3** *Bond eccentricity factor* $\psi_{ec,Na}$

**17.6.5.3.1** Modification factor for adhesive anchor groups loaded
eccentrically in tension, $\psi_{ec,Na}$, shall be calculated by Eq (17.6.5.3.1).

$$ \psi_{ec,Na} = \frac{1}{1 + \frac{e_N'}{c_{Na}}} \leq 1.0 \tag{17.6.5.3.1} $$

**17.6.5.3.2** If the loading on an adhesive anchor group is such that only
some of the adhesive anchors are in tesnion, only those adhesive anchors
that are in tension shall be considered for determining eccentricity $e_N'$
in Eq. (17.6.5.3.1) and for the calculation of $N_{ag}$ according to Eq.
(17.6.5.1b)

**17.6.5.3.3** If a load is eccentric about two orthogonal axes, $\psi_{ec,Na}$
shall be calculated for each axis individually, and the product of these
factors shall be used as $\psi_{ec,Na}$ in Eq. (17.6.5.1b).

**17.6.5.4** *Bond edge effect factor*, $\psi_{ed,Na}$

**17.6.5.4.1** Modification factor for edge effects for single adhesive
anchors or adhesive anchor groups in tension, $\psi_{ed,Na}$ shall be
determined by (a) or (b) using the critical distance $c_{Na}$ as defined in
Eq. (17.6.5.1.2b).

(a) if $c_{a,min} \geq c_{Na}$, then $$\psi_{ed,Na}=1.0 \tag{17.6.5.4.1a}$$
(b) if $c_{a,min} < c_{Na}$, then $$\psi_{ed,Na}=0.7 + 0.3 \frac{c_{a,min}}{c_{Na}} \tag{17.6.5.4.1b}$$

**17.6.5.5** *Bond splitting factor*, $\psi_{cp,Na}$

**17.6.5.5.1** Modification factor for adhesive anchors designed for
uncracked concrete in accordance with 17.6.5.1 without supplementary
reinforcement to control splitting, $\psi_{cp,Na}$, shall be determined by
(a) or (b) where $c_{ac}$ is defined in 17.9.5

(a) if $c_{a,min} \geq c_{ac}$, then $$\psi_{cp,Na}=1.0 \tag{17.6.5.5.1a}$$
(b) if $c_{a,min} < c_{ac}$, then $$\psi_{cp,Na}= \frac{c_{a,min}}{c_{ac}} \geq \frac{c_{Na}}{c_{ac}} \tag{17.6.5.5.1b}$$

**17.6.5.5.2** For all other cases, $\psi_{cp,Na}$ shall be taken as 1.0.

### 17.7 - Shear strength

**17.7.1** *Steel strength of anchors in shear*, $V_{sa}$

**17.7.1.1** Nominal steel strength of anchors in shear as governed by
steel, $V_{sa}$, shall be evaluated based on the properties of the anchor
material and the physical dimensions of the anchors. If concrete breakout
is a potential failure mode, the required steel shear strength shall be
consistent with the assumed breakout surface.

**17.7.1.2** Nominal strength of an anchor in shear, $V_sa$, shall not
exceed (a) through (c):

(a) For cast-in headed stud anchor

$$ V_{sa} = A_{se,V}f_{uta} \tag{17.7.1.2a} $$

where $A_{se,V}$ is the effective cross-sectional area of an anchor in
shear, in.<sup>2</sup> and $f_{uta}$ used for calculations shall not exceed
either $1.9f_{ya}$ or 125,000 psi.

(b) For cast-in headed bolt and hooked bolt anchors and for post-installed
anchors where sleeves do not extend through the shear plane

$$ V_{sa} = 0.6A_{se,V}f_{uta} \tag{17.7.1.2b} $$

where $A_{se,V}$ is the effective cross-sectional area of an anchor in
shear, in.<sup>2</sup>, and the value of $f_{uta}$ shall not exceed either $1.9f_{ya}$
or 125,000 psi.

(c) For post-installed anchors where sleeves extend through the shear
plane, $V_{sa}$ shall be based on the 5 percent fractile of results of
tests performed and evaluated in accordance with ACI 355.2. Alternatively,
Eq. (17.7.1.2b) shall be permitted to be used.

**17.7.1.2.1** If anchors are used with built-up grout pads, nominal
strength $V_{sa}$ calculated in accordance with 17.7.1.2 shall be
multiplied by 0.80.

**17.7.2** *Concrete Breakout strength of anchors in shear*, $V_{cb}$

**17.7.2.1** Nominal concrete breakout strength in shear, $V_{cb}$ of a
single anchor or $V_{cbg}$ of an anchor group satisfying 17.5.1.3.1, shall
be calculated in accordance with (a) through (d):

(a) For shear perpendicular to the edge on a single anchor

$$ V_{cb} = \frac{A_{Vc}}{A_{Vco}} \psi_{ed,V} \psi_{c,V} \psi_{h,V} V_b \tag{17.7.2.1a} $$

(b) For shear perpendicular to the edge on an anchor group

$$ V_{cbg} = \frac{A_{Vc}}{A_{Vco}} \psi_{ec,V} \psi_{ed,V} \psi_{c,V} \psi_{h,V} V_b \tag{17.7.2.1b} $$

(c) For shear parallel to an edge, $V_{cb}$ or $V_{cbg}$ shall be permitted
to be twice the value of the shear calculated by Eq. (17.7.2.1a) or
(17.7.2.1b), respectively, with the shear assumed to act perpendicular to
the edge and $\psi_{ed,V}$ taken equal to 1.0.

(d) For anchors located at a corner, the limiting nominal concrete breakout
strength shall be calculated for each edge, and the lesser value shall be
used.

where $\psi_{ec,V}, \psi_{ed,V}, \psi_{c,V}$ and $\psi_{h,V}$ are given in
17.7.2.3, 17.7.2.4, 17.7.2.5, and 17.7.2.6, respectively.

**17.7.2.1.1** $A_{Vc}$ is the projected area of the failure surface on the
side of the concrete member at its edge for a single anchor or an anchor
group. It shall be permitted to evaluate $A_{Vc}$ as the base of a
truncated half-pyramid projected on the side face of the member where the
top of the half-pyramid is given by the axis of the anchor row selected as
critical. The value of $c_{a1}$ shall be taken as the distacne from the
edge to this axis. $A_{Vc}$ shall not exceed $nA_{Vco}$, where $n$ is the
number of anchors in the group.

**17.7.2.1.2** If anchors are located in narrow sections of limited
thickness such that both edge distances $c_{a2}$ and thickness $h_a$ are
less than $1.5c_{a1}$, the value of $c_{a1}$ used to calculate $A_{Vc}$ in
accordance with 17.7.2.1.1 as well as for the equations in 17.7.2.1 through
17.7.2.6 shall not exceed the greatest of (a) through (c).

&nbsp;&nbsp;(a) $c_{a2}/1.5$, where $c_{a2}$ is the greatest edge distance
&nbsp;&nbsp;(b) $h_a/1.5$
&nbsp;&nbsp;(c) $s/3$, where $s$ is the maximum spacing perpendicular to
&nbsp;&nbsp;&nbsp;&nbsp;direction of shear, between anchors within a group

**17.7.2.1.3** $A_{Vco}$ is the projected area for a single anchor in a
deep member with a distance from edges of at least $1.5c_{a1}$ in the
direction perpendicular to the shear. It shall be permitted to calculate
$A_{Vco}$ by Eq. (17.7.2.1.3), which gives the area of the base of a
half-pyramid with a side length parallel to the edge of $3c_{a1}$ and a
depth of $1.5c_{a1}$.

$$ A_{Vco} = 4.5(c_{a1})^2 \tag{17.7.2.1.3} $$

**17.7.2.1.4** If anchors are located at varying distances from the edge
and the anchors are welded to the attachment so as to distribute the force
to all anchors, it shall be permitted to evaluate the strength based on the
distance to the farthest row of anchors from the edge. In this case, it
shall be permitted to base the value of $c_{a1}$ on the distance from the
edge to the axis of the farthest anchor row that is selected as critical,
and all of the shear shall be assumed to be resisted by this critical
anchor row alone.

**17.7.2.2** *Basic single anchor breakout strength*, $V_b$

**17.7.2.2.1** Basic concrete breakout strength of a single anchor in shear
in cracked concrete, $V_b$ shall not exceed the lesser of (a) and (b):

$$\text{(a)}\ V_b = \bigg(7\bigg(\frac{l_e}{d_a}\bigg)^{0.2} \sqrt{d_a}\bigg) \lambda_a \sqrt{f_c'}(c_{a1})^{1.5} \tag{17.7.2.2.1a} $$

where $l_e$ is the load-bearing length of the anchor for shear:

$l_e = h_{ef}$ for anchors with a constant stiffness over the full length
of embedded section such as headed studs and post-installed anchors with
one tubular shell over full length of the embedment depth;

$l_e = 2d_a$ for torque-controlled expansion anchors with a distance sleeve
separated from expansion sleeve;

$l_e \leq 8 d_a$ in all cases.

$$\text{(b)}\ V_b = 9 \lambda_a \sqrt{f_c'}(c_{a1})^{1.5} \tag{17.7.2.2.1b} $$

**17.7.2.2.2** For cast-in headed studs, headed bolts, or hooked bolts that
are continuously welded to steel attachments, basica concrete breakout
strength of a single anchor in shear in cracked concrete, $V_b$, shall be
the lesser of Eq. (17.7.2.2.1b) and Eq. (17.7.2.2.2) provided that (a)
through (d) are satisfied.

$$ V_b = \bigg(8\bigg(\frac{l_e}{d_a}\bigg)^{0.2} \sqrt{d_a}\bigg) \lambda_a \sqrt{f_c'}(c_{a1})^{1.5} \tag{17.7.2.2.2} $$

where $l_e$ is defined in 17.7.2.2.1.

(a) Steel attachment thickness is the greater of $0.5d_a$ and 3/8 in.
(b) Anchor spacing $s$ is at least 2.5 in.
(c) Reinforcement is provided at the corners if $c_{a2} \leq 1.5h_{ef}$
(d) For anchor groups, the strength is calculated based on the strength of
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;the row of anchors farthest from the edge.

**17.7.2.3** *Breakout eccentricity factor* $\psi_{ec,V}$

**17.7.2.3.1** Modification factor for anchor groups loaded eccentrically
in shear, $\psi_{ec,V}$, shall be calculated by Eq. (17.7.2.3.1).

$$ \psi_{ec,V} = \frac{1}{1 + \frac{e_v'}{1.5c_{a1}}} \leq 1.0 \tag{17.7.2.3.1} $$

**17.7.2.3.2** If the loading on an anchor group is such that only some of
the anchors in the group are in shear, only those anchors that are in shear
in the same direction shall be considered for determining the eccentricity
$e_V'$ in Eq. (17.7.2.3.1) and for the calculation of $V_{cbg}$ according
to Eq. (17.7.2.1b).

**17.7.2.4** *Breakout edge effect factor*, $\psi_{ed,V}$

**17.7.2.4.1** Modification factor for edge effects for single anchors or
anchor groups loaded in shear, $\psi_{ed,V}$, shall be determined by (a) or
(b) using the lesser value of $c_{a2}$.

(a) If $c_{a2} \geq 1.5c_{a1}$, then

$$ \psi_{ed,V} = 1.0 \tag{17.7.2.2.1a} $$

(b) If $c_{a2} < 1.5 c_{a1}$, then

$$ \psi_{ed,V} = 0.7 + 0.3 \frac{c_{a2}}{1.5c_{a1}} \tag{17.7.2.4.1b} $$

**17.7.2.5** *Breakout cracking factor*, $\psi_{c,V}$

**17.7.2.5.1** Modification factor for the influence of cracking in anchor
regions at service load levels and presence or absence of supplementary
reinforcement, $\psi_{c,V}$ shall be determined as follows:

(a) for anchors located in a region of a concrete member where analysis
indicates no cracking at service load levels, $\psi_{c,V}$ shall be
permitted to be 1.4.
(b) For anchors located in a region of a concrete member where analysis
indicates cracking at service load levels, $\psi_{c,V}$ shall be in
accordance with Table 17.7.2.5.1.

**Table 17.7.2.5.1 - Modification factor where analysis indicates cracking at service load levels,** $\psi_{c,V}$

| Condition                                                                                       | $\psi_{c,V}$ |
|:-----------------------------------------------------------------------------------------------:|:------------:|
| Anchors without supplementary reinforcemetn or with edge reinforcement smaller than a No. 4 bar | 1.0          |
| Anchors with reinforcement of at least a No. 4 bar or greater between the anchor and the edge   | 1.2          |
| Anchors with reinforcement of at least a No. 4 bar or greater between the anchor and the edge, and with the reinforcement enclosed within stirrups spaced at not more than 4 in.  | 1.4 |

**17.7.2.6** *Breakout thickness factor*, $\psi_{h,V}$

**17.7.2.6.1** Modification factor for anchors located in a concrete member
where $ h_a < 1.5 c_{a1}, \psi_{h,V} $ shall be calculated by Eq. (17.7.2.6.1)

$$ \psi_{h,V} = \sqrt{\frac{1.5 c_{a1}}{h_a}} \geq 1.0 \tag{17.7.2.6.1} $$



**17.7.3** *Concrete pryout strength of anchors in shear*, $V_{cp}$ *or* $V_{cpg}$

**17.7.3.1** Nominal pryout strength, $V_cp$ of a single anchor or $V_{cpg}$
of an anchor group satisfying 17.5.1.3.1, shall not exceed (a) or (b),
respectively.

(a) For a single anchor

$$ V_{cp} = k_{cp} N_{cp} \tag{17.7.3.1a} $$

(b) For an anchor group

$$ V_{cpg} = k_{cp}N_{cpg} \tag{17.7.3.1b} $$

where

$k_{cp} = 1.0$ for $h_{ef} < 2.5\ in.$
$k_{cp} = 2.0$ for $h_{ef} \geq 2.5\ in.$

**17.7.3.1.1** For cast-in anchors and post-installed expansion, screw, and
undercut anchors, $N_{cp}$ shall be taken as $N_{cb}$ calculated by Eq.
(17.6.2.1a) and for adhesive anchors, $N_{cp}$ shall be the lesser of $N_a$
calcuated by Eq. (17.6.5.1a) and $N{cb}$ calculated by Eq. (17.6.2.1a).

**17.7.3.1.2** For cast-in anchors and post-installed expansion, screw and
undercut anchors, $N_{cpg}$ shall be taken as $N_{cbg}$ calculated by Eq.
(17.6.2.1b), and for adhesive anchors, $N_{cpg}$ shall be the lesser of
$N_{ag}$ calculated by Eq. (17.6.5.1b) and $N_{cbg}$ calculated by Eq.
(17.6.2.1b).

### 17.8 - Tension and shear interaction

**17.8.1** Unless tension and shear interaction effects are considered in
accordance with 17.5.2.3, anchors or anchor groups that resist both tension
and shear shall satisfy 17.8.2 and 17.8.3. The values of $\phi N_n$ and
$\phi V_n$ shall be in accordance with 17.5.2 or 17.10.

**17.8.2** it shall be permitted to neglect the interaction between tension
and shear if (a) or (b) is satisfied.

(a)

$$ N_{ua} / (\phi N_n) \leq 0.2 \tag{17.8.2a} $$

(b)

$$ V_{ua} / (\phi V_n) \leq 0.2 \tag{17.8.2b} $$

**17.8.3** If $N_{ua}/(\phi N_n) > 0.2$ for the governing strength in
tension and $V_{ua}/(\phi V_n) \leq 0.2$ for the governing strength in
shear, then Eq. (17.8.3) shall be satisfied.

$$ \frac{N_{ua}}{\phi N_n} + \frac{V_{ua}}{\phi V_n} \leq 1.2 \tag{17.8.3} $$

## AISC Design Guide 1 - Base Connection Design for Steel Structures

### Chapter 1 — Introduction

Column base connections transfer gravity loads, lateral forces, and moments from a steel column to the concrete foundation. AISC Design Guide 1 (3rd Edition, 2024) covers three primary configurations:

- **Exposed base plate** — most common; anchor rods resist uplift and provide erection stability; shear via friction, direct bearing, or a shear lug
- **Embedded column base** — column flanges bear against surrounding concrete, resisting large moments without requiring oversized surface anchors
- **Exposed base with gusset or brace** — gusset plate transfers brace axial force into the base plate; anchor rods carry the net uplift component

For exposed bases, strength is usually governed either by **anchor rod steel** (tension or shear) or by the **ACI 318 Chapter 17 concrete breakout** limit states. Understanding which controls—and how to shift it by adjusting embedment depth, edge distances, or adding anchor reinforcement—is the primary design skill this guide develops.

### Chapter 2 — Materials

#### 2.1 Anchor Rod Grades and Selection

ASTM F1554 is the standard specification for steel construction anchor rods. Three grades cover the practical design range:

| Grade | F_y (ksi) | F_u (ksi) | Color | Ductile¹ | Max d_a (in) |
|-------|-----------|-----------|-------|------------|---------------|
| F1554 Gr 36 | 36 | 58 | Blue | Yes | 4 |
| F1554 Gr 55 | 55 | 75 | Yellow | Yes | 4 |
| F1554 Gr 105 | 105 | 125 | Red | No | 3 |

¹Ductile per ACI 318-19 Table 20.2.2.5: elongation ≥ 14%, area reduction ≥ 30%.

**Selection guidance:**
- **Gr 36 (blue)**: preferred for seismic SDC C–F weak-base designs and whenever weldability is needed; ductile without preheat
- **Gr 55 (yellow)**: stronger but still ductile; good value for larger diameter rods in moderate seismic zones
- **Gr 105 (red)**: high-strength but NOT ductile — prohibited from seismic weak-base designs; suitable only for non-seismic or strong-base applications

Field identification via **color code** is critical; misidentified rods are a known failure mode in field repairs.

Standard thread form: **UNC** (Unified National Coarse). The effective tensile stress area A_se is computed from UNC thread geometry and is the critical area for both steel strength and ACI 318 concrete checks. `AnchorBolts` selects φ based on `is_ductile`: 0.75/0.65 (tension/shear) for ductile rods vs. 0.65/0.60 for non-ductile, per ACI 318-19 Table 17.5.3(a).

```python
from civilpy.structural.concrete import AnchorBolts, _unc_stress_area

# Compare ASTM F1554 grades for a single 1-in. anchor rod
# (single anchor, far from all edges, cracked concrete)
d_a = 1.0
A_se = _unc_stress_area(d_a)
print(f'1-in. UNC anchor rod  →  A_se,N = {A_se:.4f} in²')
print()

grades = [
    ('F1554 Gr 36',   36_000,  58_000, True),
    ('F1554 Gr 55',   55_000,  75_000, True),
    ('F1554 Gr 105', 105_000, 125_000, False),
]

print(f"{'Grade':<16} {'Fy':>8} {'Fu (capped)':>12} {'phi_t':>6} "
      f"{'phi*Nsa (kip)':>14} {'phi*Vsa (kip)':>14}")
print('-' * 74)

for name, f_ya, f_uta, ductile in grades:
    rod = AnchorBolts(
        f_c=4_000, h_a=24.0, d_a=d_a, h_ef=12.0,
        f_ya=f_ya, f_uta=f_uta,
        n_x=1, n_y=1, c_a1=100.0, c_a2=100.0,
        is_cracked=True, is_ductile=ductile,
        N_ua=0, V_ua=0,
    )
    phi_N = rod.steel_tension_strength().phi_Sn / 1000
    phi_V = rod.steel_shear_strength().phi_Sn / 1000
    print(f'{name:<16} {f_ya/1000:>7.0f}k {rod.f_uta/1000:>11.0f}k '
          f'{rod.phi_steel_tension:>6.2f} {phi_N:>14.2f} {phi_V:>14.2f}')
```

#### 2.2 Base Plate Material

- **ASTM A572 Grade 50** is the standard: readily available, weldable, Fy = 50 ksi
- **ASTM A36** is acceptable for thick plates where Grade 50 is impractical or where through-thickness properties matter; Fy = 36 ksi means a thicker plate is required for the same flexure demand
- Avoid flame-cut edges with tight reentrant corners; saw-cut or machine-finished bearing surfaces are preferred under high contact pressures

#### 2.3 Grout and Concrete

The column base plate typically sits on a **non-shrink hydraulic cement grout pad** (1 to 2 in. thick placed after the column is plumbed and leveled):
- Minimum 28-day compressive strength: ≥ 5,000 psi
- Dry-pack for small plates; pumpable grout for larger bases
- After plumbing, back off leveling nuts and pressure-grout the void from one side

ACI 318-19 caps f’c at **10,000 psi** for cast-in anchor breakout (§17.3.1). Lightweight concrete reduces breakout capacity through λ_a (1.0 normalweight; 0.75 all-lightweight). A grout pad under shear reduces anchor shear capacity by 0.80 per ACI 318-19 §17.7.1.2.1 — set `grout_pad=True` in `AnchorBolts`.

### Chapter 3 — Base Connection Selection and Structural Modeling

#### 3.2 Connection Configurations

**Exposed bases — pinned:**
Four rods minimum (OSHA erection stability). Rods are typically small (3/4 or 7/8 in.) and designed for erection loads only; the column splice above provides the moment capacity.

**Exposed bases — moment:**
Anchor rods at the tension flange carry overturning uplift; the base plate cantilevers over the footing concrete and must be checked in flexure. This is the most common configuration checked with `AnchorBolts`.

**Exposed bases with gusset or brace:**
The gusset plate transfers horizontal brace force components into the base plate. For double-diagonal braces, horizontal components cancel at the base, significantly reducing anchor rod uplift demands.

**Embedded bases:**
Used when exposed-plate moment demands would require an impractically large plate or very deep anchor embedment. Column flange bearing transfers moment far more efficiently than surface anchors. Anchor rods are nominal (erection stability only). See Chapter 5.

#### 3.3 Structural Modeling of Base Fixity

The rotational stiffness assigned at the base directly affects column drift and P-Δ amplification. Modeling a flexible base as fixed is a non-conservative error.

| Connection Type | Modeling Recommendation |
|-----------------|-------------------------|
| Pinned exposed base | Pin (zero moment) |
| Weak-axis moment base plate | Elastic spring; k_θ from rod elongation + plate flex |
| Strong embedded base | Fixed or near-fixed |
| Shallow embedded base | Moderate spring per DG1 §3.5 |

When in doubt, **bracket** the analysis between pinned and fixed to identify the worst case for each design check.

### Chapter 4 — Design of Exposed Column Base Connections

#### 4.2 Overall Design Process

1. **Determine factored demands** (N_u, V_u, M_u) from LRFD load combinations at the column base
2. **Classify eccentricity**: e = M_u/N_u; if e ≤ B·N/6 the full plate is in compression (simple bearing); if e > B·N/6 some anchors are in tension
3. **Size the base plate** for concrete bearing f_p (§4.3.1) and plate flexure via cantilever lengths m, n, λn’ (§4.3.2)
4. **Compute anchor rod tension** from overturning equilibrium (elastic or plastic stress block, depending on plate stiffness assumption)
5. **Check ACI 318 Chapter 17** for all applicable concrete and steel limit states (§4.4)
6. **Add anchor reinforcement** if concrete controls — hairpin U-bars or supplementary stirrups shift the governing limit state to steel (§4.4.3)
7. **Design the weld** of column to base plate; check for eccentric shear if needed

#### 4.3 Base Plate Sizing

**Concrete bearing (§4.3.1):**

The factored bearing pressure under the plate must not exceed the ACI 318 design bearing strength:

$$f_{p,\text{actual}} = \frac{P_u}{A_1} \leq \phi_c \cdot 0.85 f'_c \sqrt{\frac{A_2}{A_1}} \leq 1.7\phi_c f'_c$$

where A_1 = base plate area, A_2 = supporting concrete plan area, and φ_c = 0.65.

**Plate flexure (§4.3.2):**

The plate cantilevers beyond the column “footprint” (0.95d × 0.80b_f). Three cantilever lengths are checked:

$$m = \frac{N - 0.95d}{2}, \quad n = \frac{B - 0.80b_f}{2}, \quad \lambda n' = \lambda\frac{\sqrt{d \cdot b_f}}{4}$$

The governing length is ℓ = max(m, n, λn’), and the minimum plate thickness under uniform bearing pressure f_p is:

$$t_{\text{min}} = \ell \sqrt{\frac{2 f_p}{0.9 F_y}}$$

Under combined loading (moment + axial) only the compression zone is in bearing; the tension zone is carried by anchor rods.

```python
import math

# AISC DG1 §4.3: size a base plate for a W10×45 column
# P_u = 150 kips (pure compression), f'c = 4,000 psi, footing 24×24 in

# W10×45 properties
d_col  = 10.10  # depth (in)
bf_col =  8.02  # flange width (in)

P_u         = 150_000   # lb
F_y_plate   = 36_000    # psi (A36)
f_c         =  4_000    # psi

# Try 14×14-in plate on 24×24-in footing
N, B = 14.0, 14.0
A1   = N * B
A2   = 24.0 * 24.0

phi_c  = 0.65
fp_max = phi_c * 0.85 * f_c * min(2.0, math.sqrt(A2 / A1))
fp_act = P_u / A1

print(f'Base plate: {N:.0f}×{B:.0f} in   A1={A1:.0f} in²  A2={A2:.0f} in²')
print(f'fp_actual = {fp_act:.0f} psi ({fp_act/1000:.2f} ksi)')
print(f'fp_max    = {fp_max:.0f} psi ({fp_max/1000:.2f} ksi)  bearing: {"OK" if fp_act <= fp_max else "NG"}')
print()

# Cantilever lengths
m  = (N - 0.95 * d_col) / 2
n  = (B - 0.80 * bf_col) / 2
X  = (4 * d_col * bf_col) / (d_col + bf_col) ** 2
lam    = min(2.0, math.sqrt(X)) / math.sqrt(1 + 1.0 / X) if X > 0 else 1.0
lam_np = lam * math.sqrt(d_col * bf_col) / 4
ell    = max(m, n, lam_np)

print(f'm = {m:.2f} in   n = {n:.2f} in   λn’ = {lam_np:.2f} in   ℓ = {ell:.2f} in')

# Minimum plate thickness
t_min = ell * math.sqrt(2 * fp_act / (0.9 * F_y_plate))
t_use = math.ceil(t_min * 4) / 4
print(f't_min = {t_min:.3f} in  →  use {t_use:.3f} in plate')
```

#### 4.4 Anchor Rod Design — ACI 318 Chapter 17 Flow

Once per-group N_ua and V_ua are established from base plate equilibrium, all ACI 318-19 Chapter 17 concrete limit states must be satisfied. `AnchorBolts.check_all()` runs the complete suite automatically.

| Limit State | Section | Key Parameter |
|-------------|---------|---------------|
| Steel tension | 17.6.1 | A_se,N; F_u capped at min(1.9·Fy, 125 ksi) |
| Concrete breakout (tension) | 17.6.2 | h_ef (∝ h_ef^1.5); A_Nc; edge factors |
| Pullout — headed rod | 17.6.3.2.2a | A_brg (head or washer plate bearing area) |
| Pullout — hooked rod | 17.6.3.2.2b | e_h (hook length; 3d_a ≤ e_h ≤ 4.5d_a) |
| Side-face blowout | 17.6.4 | Deep anchor near free edge (h_ef > 2.5·c_a1) |
| Steel shear | 17.7.1 | A_se,V; grout pad reduces 0.80 |
| Concrete breakout (shear) | 17.7.2 | c_a1 (edge in shear direction) |
| Pryout | 17.7.3 | k_cp × N_cbg (governs for shallow anchors) |
| Tension-shear interaction | 17.8 | Unity check Eq. 17.8.3 |

**SDC C–F seismic reduction:** `_phi_seismic = 0.75` is multiplied onto all concrete failure-mode capacities per ACI 318-19 §17.10.5.3, applied automatically when `sdc` is “C”, “D”, “E”, or “F”.

**Anchor reinforcement** (`anchor_reinf_tension` / `anchor_reinf_shear` in lb): providing properly developed hairpin U-bars per ACI 318-19 §17.5.2.1 replaces the concrete breakout check entirely, shifting the governing limit state to steel. This is the most cost-effective fix when breakout governs.

### Field Repair of Damaged Anchor Rods — Coupling Nut Extension

When an anchor rod is damaged or cut too short during construction, AISC Design Guide 1
permits extension with a **heavy-hex coupling nut** and a new concrete repour (weak-base detail).

**Critical requirement (ACI 318-19 Section 17.10.5.3 / AISC DG1):**
The coupling nut must be **frictionally isolated** from the new concrete pour — typically with
**polyethylene tape** or a **PVC sleeve** — to preserve the anchor's stretch length and maintain
ductile behavior under seismic loading.

If the coupler is *not* isolated, it acts as a bearing surface and the effective embedment depth is
reduced to the depth below the coupler:

500h_{ef,eff} = h_{ef} - L_{coupler}500

**Minimum stretch length requirement** (SDC C–F): {stretch} \geq 8 d_a$

The  object handles both scenarios:
-  reduces  for breakout calculations when isolation is absent.
-  flags SDC C–F designs where the isolated length is < 8·d_a.

```python
import math

d_a   = 1.25          # 1-1/4 in rod
h_ef  = 30.0          # original embedment
L_cpl = 6.0           # 6-in coupling nut

print("=== CASE A: Isolated coupler (correct installation) ===")
ab_isolated = AnchorBolts.from_circular(
    n_bolts=8, bolt_circle_radius=9.0,
    f_c=4_000, h_a=18.0, d_a=d_a, h_ef=h_ef,
    f_ya=55_000, f_uta=75_000,
    sdc="C",
    N_ua=9_722,   # same per-bolt demand as luminaire example
)
print(f"  h_ef used for breakout = {ab_isolated.h_ef:.1f} in  (coupler isolated — full depth)")
sl = ab_isolated.stretch_length_check()
print(f"  Stretch length: demand={sl.demand:.1f} in >= limit={sl.phi_Sn:.1f} in → {'OK' if sl.ok else 'NG'}")

print()
print("=== CASE B: Non-isolated coupler (construction error) ===")
ab_nonisolated = AnchorBolts.from_circular(
    n_bolts=8, bolt_circle_radius=9.0,
    f_c=4_000, h_a=18.0, d_a=d_a, h_ef=h_ef,
    f_ya=55_000, f_uta=75_000,
    sdc="C",
    N_ua=9_722,
    coupler_depth=L_cpl,  # coupler acts as bearing → reduces h_ef
)
print(f"  h_ef used for breakout = {ab_nonisolated.h_ef:.1f} in  (h_ef reduced by {L_cpl} in)")
print()
print("  Full summary:")
print(ab_nonisolated.summary())
```

### Chapter 5 — Embedded Column Base Connections

#### 5.1 Overview

Embedded connections resist applied moment through **horizontal bearing** of the column flanges against the surrounding concrete. This mechanism is far more efficient than surface anchor rods for large moments — the column flange becomes a load-spreading surface rather than a collection of small point loads.

Two configurations:
- **Type I** — column embedded in a footing or grade beam pocket; no vertical foundation stress transfers at the base plate level (axial load goes into the footing through the column itself or base plate bearing at the bottom)
- **Type II** — column embedded in a grade beam or wall; axial friction and vertical bearing interact with the horizontal bearing moment mechanism

Anchor rods in embedded connections are sized for **erection stability only** (OSHA four-rod minimum). They do not carry service or strength-level moment.

#### 5.2 Horizontal Bearing Mechanism

The applied moment is resisted by two equal-and-opposite bearing force resultants — one near the top of the embedded zone, one near the bottom — acting on opposite column flanges (or face bearing plates welded to the flanges). Under a simplified uniform bearing stress block over half the embedment depth:

$$M_{HB} = \phi_c \cdot f_{b,\text{lim}} \cdot b_{\text{eff}} \cdot \frac{d_e^2}{4}$$

where:
- $f_{b,\text{lim}} = \min\!\left(0.85\sqrt{A_2/A_1}\,f'_c,\; 1.7f'_c\right)$ — ACI 318-19 Eq. 22.8.3.2; maximum = 1.7f’c when A_2/A_1 ≥ 4
- φ_c = 0.65 (bearing)
- b_eff = column flange width + face bearing plate projections
- d_e = effective embedment depth

**Face bearing plates (FBPs)** are short stiffener plates welded across the column flanges. They increase b_eff, improve bearing stress distribution, and are typically required when the bare-flange capacity is insufficient.

#### 5.3 Example 5.3-1: Embedded Base for a W14×176 Column (AISC DG1)

**Given:**
- Column: W14×176 (d = 15.22 in, b_f = 15.65 in)
- Factored demands: P_u = 250 kips (compression), V_u = 96 kips, M_u = 700 kip-ft
- Concrete: f’c = 4,000 psi; large footing (A_2/A_1 ≥ 4, bearing at limit)

**Determine:** Required embedment depth with face bearing plates; check bearing capacity. Verify erection anchor shear.

The bearing stress limit at A_2/A_1 = 4: 0.85·√4·f’c = 1.7f’c = 6,800 psi. Design value: φ·f_b = 0.65 × 6,800 = 4,420 psi.

Erection rods (4 × 3/4-in. F1554 Gr 36) are not designed for moment; only the erection shear is checked against steel shear strength.

```python
import math
from civilpy.structural.concrete import AnchorBolts

# ---- Geometry and demands ------------------------------------------
bf_col  = 15.65   # W14x176 flange width (in)

M_u_kft = 700.0                 # kip-ft demand
M_u     = M_u_kft * 12          # kip-in
V_u     = 96.0                  # kips (carried by embedded flange mechanism)

# ---- ACI 318 bearing limit (max A2/A1 case) ------------------------
f_c    = 4_000    # psi
phi_c  = 0.65
f_blim = 0.85 * math.sqrt(4.0) * f_c    # = 1.7*fc = 6800 psi (unfactored)
phi_fb = phi_c * f_blim                  # = 4420 psi (design)

print(f"Concrete f'c = {f_c} psi")
print(f"f_b,lim (unfactored) = {f_blim:.0f} psi = {f_blim/1000:.2f} ksi  "
      f"(ACI 318-19 Eq. 22.8.3.2 max, A2/A1 >= 4)")
print(f"phi*f_b (design)     = {phi_fb:.0f} psi = {phi_fb/1000:.2f} ksi")
print()

# ---- Try d_e = 24 in with 3-in face bearing plates each side -------
d_e   = 24.0    # embedment depth (in)
b_fbp =  3.0    # FBP projection beyond each flange tip (in)
b_eff = bf_col + 2 * b_fbp

# Moment capacity (simplified uniform bearing block, half depth each end)
M_cap = (phi_fb / 1000) * b_eff * d_e**2 / 4    # kip-in

print(f"Try d_e = {d_e:.0f} in with {b_fbp:.0f}-in FBPs each side")
print(f"b_eff = {b_eff:.2f} in  (bf={bf_col} + 2*{b_fbp} in FBP)")
print(f"M_HB  = phi*f_b * b_eff * d_e^2/4")
print(f"      = {phi_fb/1000:.3f} kip/in^2 * {b_eff:.2f} in * {d_e**2/4:.0f} in^2")
print(f"      = {M_cap:.0f} kip-in = {M_cap/12:.0f} kip-ft")
print(f"M_u   = {M_u:.0f} kip-in = {M_u_kft:.0f} kip-ft   "
      f"DCR = {M_u/M_cap:.3f}  {'OK' if M_u <= M_cap else 'NG'}")
print()

# ---- Face bearing plate thickness check ----------------------------
# FBP cantilevers b_fbp under distributed bearing pressure phi_fb
# Required: M_fbp = phi_fb*b_fbp^2/2; Capacity: t^2*Fy/6
F_y_fbp = 50_000    # psi (A572 Gr 50 FBP)
t_fbp = b_fbp * math.sqrt(3 * phi_fb / F_y_fbp)
t_use = math.ceil(t_fbp * 4) / 4
print(f"Face bearing plate: cantilever = {b_fbp:.1f} in, Fy = {F_y_fbp/1000:.0f} ksi")
print(f"t_fbp,min = {t_fbp:.3f} in  ->  use {t_use:.3f} in plate")
print()

# ---- Erection anchor rods (AnchorBolts, tension check only) --------
# In embedded connections, all design-level shear and moment is resisted by
# the embedded column flange bearing mechanism -- NOT by anchor rods.
# Anchor rods (OSHA minimum: 4 rods) provide erection stability only.
# Check: 3/4-in F1554 Gr 36 rods for nominal erection uplift.
print('Erection rods: 4 x 3/4-in F1554 Gr 36 (erection stability, not strength design)')
erect = AnchorBolts(
    f_c=f_c, h_a=d_e, d_a=0.75, h_ef=12.0,
    f_ya=36_000, f_uta=58_000,
    n_x=1, n_y=1, c_a1=18.0, c_a2=18.0,
    is_cracked=False, has_supp_reinf=False,
    N_ua=500,    # 0.5 kips -- nominal erection uplift per rod (lb)
    V_ua=0,
)
st = erect.steel_tension_strength()
bt = erect.breakout_strength_tension()
print(f'  phi*N_sa  = {st.phi_Sn/1000:.1f} kips (steel)        '
      f'vs N_ua = {st.demand/1000:.2f} kips/rod  -> {"OK" if st.ok else "NG"}')
print(f'  phi*N_cbg = {bt.phi_Sn/1000:.1f} kips (breakout, c=18 in) '
      f'vs N_ua = {bt.demand/1000:.2f} kips/rod  -> {"OK" if bt.ok else "NG"}')
```

### Chapter 6 — Seismic Column Base Design

#### 6.1 Strong-Base vs. Weak-Base Design

In seismic design, the engineer must choose where plastic hinging occurs:

**Strong-base design:** The base connection remains elastic; the plastic hinge forms in the column above the base plate. Anchor rods and base plate are designed for the column’s plastic moment capacity, amplified by overstrength:

$$M_u^* = 1.1 R_y F_y Z_x$$

This protects the foundation from inelastic demand but requires oversized base plate components. All ACI 318 Chapter 17 concrete checks still apply.

**Weak-base design (common in highway sign, luminaire, and light pole structures):** Controlled yielding occurs in the anchor rods. This requires:
- **Ductile rods** (F1554 Gr 36 or Gr 55) — elongation ≥ 14%, reduction ≥ 30%
- **Upset thread (UT) detail** — the rod shank below the nut is machined to a reduced smooth cross-section that acts as the intended plastic fuse; the upset (enlarged) threads bear on the nut over a short length while the smooth zone undergoes uniform elongation
- **Frictional isolation** of the stretch zone from the concrete (polyethylene tape or PVC sleeve) so the rod can elongate without bond restraint
- **Stretch length ≥ 8·d_a** (ACI 318-19 §17.10.5.3) — ensures sufficient plastic elongation capacity before fracture

#### 6.2 ACI 318 Requirements for SDC C–F (§17.10)

- Concrete assumed **cracked** for all anchor limit states (§17.10.5.4)
- Post-installed anchors must be ACI 355.2 (mechanical) or 355.4 (adhesive) seismic-qualified (§17.10.3)
- All concrete failure-mode capacities multiplied by **φ_seismic = 0.75** (§17.10.5.3) — applied automatically in `AnchorBolts` when `sdc='C'` through `'F'`
- Anchor reinforcement (deformed bars, §20.2.2) can replace concrete breakout when properly developed, but bars must clear the φ-seismic reduction

#### 6.3 Stretch Length and Field Repair Considerations

In weak-base designs the isolated stretch zone is the ductile fuse:
- **New construction (UT detail):** the frictional isolation is built in during initial construction; PE tape or sleeve wraps the smooth shank before concrete pour
- **Field repair (coupling nut):** a heavy-hex coupler extends a short or damaged rod via a concrete repour; the coupler itself must be frictionally isolated from the new concrete to preserve the stretch length (see earlier Coupling Nut demo)
- The minimum stretch length 8·d_a is checked with `AnchorBolts.stretch_length_check()`

#### 6.4 Example 6.4-1: Weak-Base Seismic Design for a W12×96 Column (AISC DG1)

**Given:**
- Column: W12×96
- SDC C–F; design for overstrength: M_u = 3,600 kip-in, P_u = 376 kips (compression, gravity-dominated), V_u = 50 kips
- 4 × 1-in. ASTM F1554 Gr 55 anchor rods (ductile, upset thread detail) in a rectangular 12×12-in. pattern
- f’c = 5,000 psi; h_ef = 24 in; A_brg = 3.21 in² per rod (washer plates)
- Per-bolt demands from elastic group analysis: N_ua = 48.3 kips, V_ua = 50/4 = 12.5 kips

The seismic reduction (φ_seismic = 0.75) applies to all concrete limit states. The stretch length check with `sdc='C'` confirms the UT detail satisfies the 8·d_a = 8-in. minimum.

```python
from civilpy.structural.concrete import AnchorBolts, _unc_stress_area

# ---- Per-bolt demands (from DG1 elastic anchor group analysis) ------
N_ua_bolt = 48_300     # lb (governs at overstrength seismic load)
V_ua_bolt = 50_000 / 4  # lb (4 rods share base shear equally)

d_a  = 1.0     # nominal rod diameter (in) -- F1554 Gr 55
h_ef = 24.0    # embedment depth (in)
f_c  = 5_000   # psi

print(f'Per-bolt demands:  N_ua = {N_ua_bolt/1000:.1f} kips,  '
      f'V_ua = {V_ua_bolt/1000:.1f} kips')
print(f'h_ef = {h_ef:.0f} in,  f\'c = {f_c} psi,  SDC C (phi_seismic = 0.75)')
print()

ex641 = AnchorBolts(
    f_c=f_c,
    h_a=48.0,            # deep footing, no thickness reduction
    d_a=d_a,
    h_ef=h_ef,
    f_ya=55_000,         # F1554 Gr 55
    f_uta=75_000,
    n_x=2, n_y=2,        # 4 rods in 12x12-in rectangular pattern
    s_x=12.0, s_y=12.0,
    c_a1=100.0, c_a2=100.0,  # interior footing -- no edge effects
    A_se_N=_unc_stress_area(d_a),
    A_brg=3.21,          # washer plate bearing area per rod (in^2)
    is_cracked=True,     # ACI 318-19 Sec 17.10.5.4 -- SDC C-F always cracked
    has_supp_reinf=False,
    is_ductile=True,     # F1554 Gr 55 qualifies
    sdc='C',
    N_ua=N_ua_bolt,
    V_ua=V_ua_bolt,
)

print(ex641.summary())

# ---- Stretch length check (ACI 318-19 Sec 17.10.5.3) ---------------
sl = ex641.stretch_length_check()
L_min = 8.0 * d_a
print()
print('Stretch length check (UT detail -- PE tape isolates rod from concrete):')
print(f'  Required  8*d_a = {L_min:.1f} in')
print(f'  Available h_ef  = {h_ef:.1f} in  '
      f'(conservative proxy; actual isolated zone = from base plate to bottom nut)')
print(f'  {sl}')

# ---- Illustrate seismic penalty on concrete breakout ----------------
print()
print('Effect of SDC C seismic reduction (phi_seismic = 0.75):')
ex641_noseis = AnchorBolts(
    f_c=f_c, h_a=48.0, d_a=d_a, h_ef=h_ef,
    f_ya=55_000, f_uta=75_000,
    n_x=2, n_y=2, s_x=12.0, s_y=12.0,
    c_a1=100.0, c_a2=100.0,
    A_se_N=_unc_stress_area(d_a), A_brg=3.21,
    is_cracked=True, is_ductile=True,
    sdc='A',           # no seismic reduction -- for comparison only
    N_ua=N_ua_bolt, V_ua=V_ua_bolt,
)
bt_seis   = ex641.breakout_strength_tension()
bt_noseis = ex641_noseis.breakout_strength_tension()
print(f'  phi*N_cbg without seismic: {bt_noseis.phi_Sn/1000:.1f} kips')
print(f'  phi*N_cbg with SDC C:      {bt_seis.phi_Sn/1000:.1f} kips  '
      f'(ratio = {bt_seis.phi_Sn/bt_noseis.phi_Sn:.2f})')
```

## Advanced Shear — Shear Lugs (`ShearLugCheck`)

Anchor rods transfer shear by bending — a relatively inefficient mechanism that becomes impractical for large base shears (braced frames, wind towers, bridge piers). A **shear lug** is a short steel plate or bar welded to the underside of the base plate and embedded directly in the concrete, transferring shear through direct bearing rather than rod bending. ACI 318-19 Section 17.11 governs two limit states:

1. **Bearing strength** (§17.11.2.1): `V_brg,sl = n_sl × 1.7 × λ_a × f'c × A_ef,sl`
2. **Concrete breakout** (§17.11.3): projection method, same form as anchor rod breakout

**Scenario:** A braced-frame column base with V_u = 220 kips. Four 1-in. Gr 55 anchor rods provide only ~164 kips of shear capacity — a 25% shortfall. A single 8-in. × 6-in. shear lug (A_ef = 48 in²) solves it.

```python
from civilpy.structural.concrete import AnchorBolts, ShearLugCheck

V_u = 220_000   # lb — braced-frame base shear
f_c = 4_000

# ---- Step 1: show anchor rods alone are insufficient -------------------
rods_only = AnchorBolts(
    f_c=f_c, h_a=36.0, d_a=1.0, h_ef=15.0,
    f_ya=55_000, f_uta=75_000,
    n_x=2, n_y=2, s_x=8.0, s_y=8.0,
    c_a1=100.0, c_a2=100.0,
    is_cracked=True, is_ductile=True,
    V_ua=V_u,
)
vst = rods_only.steel_shear_strength()
print('4 x 1-in F1554 Gr 55 anchor rods (shear only):')
print(f'  phi*V_sa = {vst.phi_Sn/1000:.1f} kips  vs  V_u = {V_u/1000:.0f} kips  -> '
      f'{"OK" if vst.ok else "NG -- add a shear lug"}')
print()

# ---- Step 2: single shear lug resolves the deficiency ------------------
# 8-in wide x 6-in tall lug plate welded to base plate underside
# A_ef,sl = bearing area (w x h, minus grout pocket area) = 8 * 6 = 48 in^2
# h_sl = 3 in height above concrete surface (inside grout pocket)
lug = ShearLugCheck(
    f_c=f_c,
    h_a=36.0,
    n_sl=1,
    A_ef_sl=48.0,    # 8 in wide * 6 in tall (in^2)
    h_sl=3.0,        # lug height above grout bed (in)
    c_a1=float('inf'),  # no near edge in shear direction
    c_a2=float('inf'),
    has_supp_reinf=False,
    V_ua=V_u,
)

print('Shear lug check:')
print(lug.summary())
brg = lug.bearing_strength()
print(f'Lug bearng area: {lug.A_ef_sl:.0f} in^2,  '
      f'V_brg,sl (nominal) = {brg.V_n/1000:.1f} kips,  '
      f'phi*V_brg = {brg.phi_Sn/1000:.1f} kips')
```

## Corner Shear Breakout & Parallel Shear Direction

ACI 318-19 §17.7.2.1 covers three special cases for anchor groups near edges:

- **Perpendicular shear near two edges (corner):** when `c_a2 < 1.5·c_a1`, `AnchorBolts` automatically checks breakout from both edges and uses the minimum capacity (ACI §17.7.2.1(d))
- **Shear parallel to a free edge:** `shear_direction='parallel'` applies ACI §17.7.2.1(c), which doubles the perpendicular-breakout capacity and sets ψ_ed,V = 1.0 (the edge is not the failure plane)

**Setup:** 2×2 group of 3/4-in. Gr 36 rods near a pier corner: c_a1 = 10 in (shear direction), c_a2 = 12 in (perpendicular). V_u = 15 kips.

```python
from civilpy.structural.concrete import AnchorBolts

shared = dict(
    f_c=4_000, h_a=18.0, d_a=0.75, h_ef=8.0,
    f_ya=36_000, f_uta=58_000,
    n_x=2, n_y=2, s_x=6.0, s_y=6.0,
    c_a1=10.0, c_a2=12.0,
    is_cracked=True, has_supp_reinf=False,
    V_ua=15_000,
)

# ---- Case A: perpendicular shear near corner ---------------------------
perp = AnchorBolts(**shared, shear_direction='perpendicular')
vb_perp = perp.breakout_strength_shear()
print('Case A: shear_direction="perpendicular" (corner — checks both edges)')
print(f'  phi*V_cbg = {vb_perp.phi_Sn/1000:.2f} kips   ref: {vb_perp.reference}')
print(f'  DCR = {vb_perp.dcr:.3f}  {"OK" if vb_perp.ok else "NG"}')
print()

# ---- Case B: shear parallel to the near edge ---------------------------
par = AnchorBolts(**shared, shear_direction='parallel')
vb_par = par.breakout_strength_shear()
print('Case B: shear_direction="parallel"  (ACI 17.7.2.1(c) -- 2x perpendicular capacity)')
print(f'  phi*V_cbg = {vb_par.phi_Sn/1000:.2f} kips   ref: {vb_par.reference}')
print(f'  DCR = {vb_par.dcr:.3f}  {"OK" if vb_par.ok else "NG"}')
print()

print(f'Capacity ratio (parallel / perpendicular) = '
      f'{vb_par.phi_Sn / vb_perp.phi_Sn:.2f}  (ideally ~2.0 for no corner effect)')
```

## Post-Installed Adhesive Anchors (ACI 318-19 §17.6.5)

Adhesive anchors set `anchor_type='adhesive'` and require two additional parameters:
- `tau_uncr` — characteristic bond stress in **uncracked** concrete (psi), from the ACI 355.4 evaluation report for the specific adhesive product
- `tau_cr` — characteristic bond stress in **cracked** concrete (psi)

The bond check adds `N_ba / N_ag` (ACI §17.6.5.1.1) alongside the standard breakout checks. For post-installed anchors in **uncracked** concrete, `ψ_c,N = 1.40` (vs 1.25 for cast-in) and `ψ_cp,N` applies the splitting correction based on `c_a_min / c_ac`.

**Scenario:** Single 3/4-in. adhesive anchor, τ_uncr = 1,800 psi, uncracked concrete, f'c = 6,000 psi, h_ef = 6 in.

```python
from civilpy.structural.concrete import AnchorBolts

adhesive = AnchorBolts(
    f_c=6_000,
    h_a=24.0,
    d_a=0.75,
    h_ef=6.0,
    f_ya=55_000,
    f_uta=75_000,
    n_x=1, n_y=1,
    c_a1=8.0, c_a2=8.0,     # moderately close to edge
    is_cracked=False,        # uncracked -> psi_c,N = 1.40
    has_supp_reinf=False,
    anchor_type='adhesive',
    tau_uncr=1_800,          # psi -- from ACI 355.4 evaluation report
    N_ua=8_000,
    V_ua=0,
)

print('Post-installed adhesive anchor — full limit-state check:')
print(adhesive.summary())
print()

# Show key intermediate values
bond = adhesive.bond_strength_tension()
bt   = adhesive.breakout_strength_tension()
print(f'psi_c,N  = {adhesive._psi_c_N():.2f}  '
      f'(1.40 for post-installed uncracked, ACI 17.6.2.5.1)')
print(f'psi_cp,N = {adhesive._psi_cp_N():.3f}  '
      f'(splitting factor: c_a_min/c_ac = {adhesive.c_a_min:.1f}/{adhesive.c_ac:.1f})')
print()
print(f'Bond strength phi*N_ag = {bond.phi_Sn/1000:.2f} kips  '
      f'(ACI 17.6.5 -- bond governs when < breakout)')
print(f'Breakout   phi*N_cbg   = {bt.phi_Sn/1000:.2f} kips')
print(f'Governing:             = {min(bond.phi_Sn, bt.phi_Sn)/1000:.2f} kips  '
      f'({"bond" if bond.phi_Sn < bt.phi_Sn else "breakout"})')
```

## The Bailout Feature: Anchor Reinforcement Override

When concrete breakout governs and limits anchor capacity, **anchor reinforcement** (hairpin U-bars or ties) per ACI 318-19 §17.5.2.1 can replace the concrete breakout check entirely. `AnchorBolts` accepts:
- `anchor_reinf_tension` (lb) — design strength φR_n of the reinforcement in tension; replaces concrete breakout in the tension table
- `anchor_reinf_shear` (lb) — same for shear direction

**Scenario:** 2×2 cast-in group on a **narrow pier** (c_a1 = c_a2 = 8 in, s_x = s_y = 4 in). N_ua = 60 kips. Without reinforcement, concrete breakout fails badly. With 4 properly developed #5 hairpin bars (φR_n ≈ 0.9 × 31 ksi × 4 × 0.31 in² ≈ 34.6 kips... use 45 kips for hairpins with seismic hooks), the governing limit state shifts to steel.

```python
from civilpy.structural.concrete import AnchorBolts

base_kwargs = dict(
    f_c=4_000, h_a=18.0, d_a=1.0, h_ef=12.0,
    f_ya=36_000, f_uta=58_000,
    n_x=2, n_y=2, s_x=4.0, s_y=4.0,
    c_a1=8.0, c_a2=8.0,    # narrow pier -- tight edge distances
    is_cracked=True, has_supp_reinf=False,
    N_ua=60_000,
    V_ua=0,
)

# ---- Without anchor reinforcement (breakout governs and fails) ----------
no_reinf = AnchorBolts(**base_kwargs)
bt_fail = no_reinf.breakout_strength_tension()
print('WITHOUT anchor reinforcement:')
print(no_reinf.summary())
print(f'  Concrete breakout DCR = {bt_fail.dcr:.3f}  '
      f'{"OK" if bt_fail.ok else "NG -- reinforcement needed"}')
print()

# ---- With anchor reinforcement (breakout check replaced) ----------------
# 4 x #5 hairpin U-bars, properly developed on both sides of potential
# failure plane: phi*R_n = 0.9 * 60 ksi * 4 * 0.31 in^2 = 66.96 kips
phi_Rn_reinf = 0.9 * 60_000 * 4 * 0.31   # lb
with_reinf = AnchorBolts(**base_kwargs, anchor_reinf_tension=phi_Rn_reinf)
print(f'WITH anchor reinforcement (4 x #5 hairpins, phi*R_n = {phi_Rn_reinf/1000:.1f} kips):')
print(with_reinf.summary())
print(f'  Governing tension limit state: '
      f'{with_reinf.governing_tension_strength().label}')
```

## CAD Integration — DXF Importer (`AnchorBolts.from_dxf`)

`AnchorBolts.from_dxf()` builds an instance by reading bolt and concrete boundary geometry directly from a DXF file. Two CAD layers are required:

| Layer | Entity type | Meaning |
|-------|------------|----------------------------------------------|
| `BOLTS` | `CIRCLE` | Each circle center = bolt location; radius = d_a/2 |
| `CONCRETE` | `LWPOLYLINE` | Closed polyline = concrete boundary; used to derive c_a1, c_a2 |

The importer auto-detects **circular** patterns (all bolts within 10% of the same radius from the group centroid → delegates to `from_circular()`) vs. **rectangular** grids (reconstructs n_x, n_y, s_x, s_y from sorted bolt coordinates).

Material properties not derivable from geometry (f_c, h_a, h_ef, f_ya, f_uta) are passed as keyword arguments along with any other `AnchorBolts` parameters (N_ua, V_ua, sdc, ...).

## Visualization — Anchor Group Plot (`.plot()`)

`AnchorBolts.plot()` renders a 2D plan-view diagram showing:
- Concrete boundary (grey filled rectangle)
- Bolt locations (blue circles at correct scale)
- 1.5·h_ef tension breakout cone (dashed red circle/boundary)
- Shear demand arrow (orange) and tension label (green ⊕)

**Showcase:** A 3×2 group with asymmetric spacing near a corner (c_a1 = 12 in, c_a2 = 8 in) under combined tension and shear. The plot makes it visually clear how the breakout cone interacts with the free edges and why the corner governs the shear check.

```python
import matplotlib
matplotlib.use('Agg')   # non-interactive backend for notebook environments
import matplotlib.pyplot as plt
from civilpy.structural.concrete import AnchorBolts

# Asymmetric 3x2 group near a corner
showcase = AnchorBolts(
    f_c=4_000,
    h_a=24.0,
    d_a=0.875,
    h_ef=10.0,
    f_ya=36_000,
    f_uta=58_000,
    n_x=3, n_y=2,
    s_x=5.0, s_y=7.0,     # different spacing in each direction
    c_a1=12.0,             # near edge in shear direction
    c_a2=8.0,              # near edge perpendicular (corner condition)
    is_cracked=True,
    has_supp_reinf=False,
    shear_direction='perpendicular',
    N_ua=30_000,           # lb tension
    V_ua=12_000,           # lb shear toward edge
)

print(showcase.summary())
print()

fig, ax = showcase.plot(title='3x2 Group Near Corner -- Plan View')
plt.savefig('anchor_plot_showcase.png', dpi=120, bbox_inches='tight')
plt.show()
print('Plot saved to anchor_plot_showcase.png')
```

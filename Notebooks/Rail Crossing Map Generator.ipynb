```python
from civilpy.transportation.fra_xing import find_nearest_crossing, create_crossing_map
```

```python
# --- Configuration ---
# Set the coordinates for your search
snow_road_lon = -81.693553
snow_road_lat = 41.499771

# Set the railroad you want to highlight
target_railroad_to_find = 'NS'

# --- Run the search and create the map ---
all_features, nearest = find_nearest_crossing(snow_road_lon, snow_road_lat)

if nearest:
    attrs = nearest.get('attributes', {})
    geom = nearest.get('geometry', {})
    print("\n--- Nearest Grade Crossing Found ---")
    print(f"  Crossing ID: {attrs.get('CROSSING')}")
    print(f"  Location: {attrs.get('STREET', 'N/A')} in {attrs.get('CITYNAME', 'N/A')}, {attrs.get('STATENAME', 'N/A')}")
    print(f"  Railroad: {attrs.get('RAILROAD')}")
    print(f"  Coordinates: ({geom.get('y')}, {geom.get('x')})")
    print("------------------------------------")
    
    create_crossing_map(snow_road_lat, snow_road_lon, all_features, nearest, target_railroad_to_find)
```

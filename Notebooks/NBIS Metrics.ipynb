# NBIS Assessment Metrics (Process)

FHWA developed a series of metrics they utilize to assess bridge inspection  
programs accross the country.

This document was intended to develop a system of working set of these  
metric calculations implemented in the python programming language.

Each metric consists of the following six parts:

1. NBIS component to be reviewed (NBIS Reference)
2. Evaluation Criteria (Criteria)
3. Population
4. Compliance Levels
5. Assessment Levels
6. Commentary

Each metric is assessed into one of four levels of compliance:

**Compliant**: Adheres to the NBIS regulation  
**Substantially Compliant**: Adheres to the regulation with minor deficiencies  
that do not adversly affect the overall effectiveness of the program and  
are isolated in nature  
**Non-Compliant**: Not adhering to the NBIS regulation, thus failing to meet  
one or more of the Substantial Compliance criteria for a metric. Identified  
deficiencies may adversely affect the overall effectiveness of the program.  
**Conditionally Compliant**: Taking corrective action in conformance with an  
FHWA-approved PCA to achieve compliance with the NBIS.

Each metric is assessed annually at either the Minimum (Min-AL),  
Intermediate (Int-AL), or In-Depth (InD-AL) assessment level.

See FHWA for exact guidance on which is performed when, basically it's  
FHWA's discretion.

FHWA will provide a report summarizing their review.

## Assessment Levels

3 levels of assessments,

- **Minimum Assessment Level (Min-AL)** - A review based on information from  
past assessments and the reviewer's knowledge of the current practice as it  
relates to the metric. This is basically just a review of existing knowledge  
to make sure an intermediate assessment isn't necessary.

- **Intermediate Assessment Level (Int-AL)** - In addition to the methods in  
Min-AL, the reviewer uses more intense methods like random sampling of  
inspection records, analysis of bridge inventory data, site visits, interviews  
and review of documentation to determine compliance.  

    For metrics that utilize random sampling, Int-AL involves Tier 1 random  
    sampling using a margin of error (MOE) of 15% and a level of confidence  
    (LOC) 80% to review bridge records. A Tier 2 random sampling, utilizing an  
    MOE of 10% and an LOC of 80% is used when the results of the Tier 1 sample  
    are inconclusive. Random samples are selected from the population  
    identified for the specific metric. 

- **In-Depth Assessment Level (InD-AL)** - This level is the most intensive  
and is used to supplement the methods described in Int-AL with larger  
random sample sizes, more interviews, or research of records, documentation,  
and/or history. There are two ways to perform InD-AL: Division and  
nationally directed. For Division InD-AL reviews, the reviewer develops  
guidelines in addition to the Int-AL guidelines as appropriate for the  
metric or issue being assessed, with concurrence from the BSE, and conducts  
the review in accordance with guidelines.

## Metric Assessment Reports (MARs)

```python
def get_required_samples(population, assessment_type, tier):
    """
    Calculates sample size treating table values as "Starting Thresholds".
    """
    cutoffs = LOOKUP_TABLE.get((assessment_type, tier))

    if cutoffs is None:
        raise ValueError(f"Invalid condition: {assessment_type} {tier}")

    index = bisect.bisect_right(cutoffs, population)

    sample_size = index

    return min(sample_size, len(cutoffs))
```

nominal axial strength, $ P_{nl} $ for local buckling given by:

$$ P_{nl} = A_e F_n $$

Where:

$ F_n $ = Global column stress

### Effective Area, $ A_e $

- Section E3.1.1 - Members without holes
    - States that the effective width of all elements shall be found according  
to Appendix 1 of AISI S100-16 at stress $F_n$
- Appendix 1 organizes the *Effective Width* of Elements into multiple criteria
- Applicable Sections:
    - Section 1.1 Effective Width of Uniformly Compressed Stiffened Elements
          - Web
    - Section 1.2 Effective Width of Unstiffened Elements
          - Flanges

#### Section 1.1 - *Efective Width of Uniformly Compressed Stiffened Elements*

- Effective Width $ b_e $ is defined as:
    - $ b_e = \rho w $

Where:

$ w $ = Flat Width as displayed in Figure 1.1-1  
$ \rho $ = Local Reduction Factor



Local Reduction Factor: $\rho$

- When Slenderness Factor, $ \lambda \leq 0.673 $:
    - $ \rho $ = 1.0
- When $ \lambda \ge 0.673 $:

$$ \rho = \frac{1 - \frac{0.22}{\lambda}}{\lambda} $$

Where:

$ \lambda = \sqrt{\frac{f}{f_{crl}}} $ 

For compression memebrs, $ f = F_n $ as determined in accordance with  
Chapter E in AISI S100-16.

$$ F_{crl} = \frac{k \pi^2 E}{12(1-v^2)}\bigg(\frac{t}{w}\bigg)^2 $$

Where:

$ k $ = Plate Buckling Coefficient (Table C-1-1)  
$ E $ = Youngs Modulus  
$ v $ = Poisson's Ratio  
$ t $ = Thickness of uniformly compressed stiffened element  
$ w $ = Flat Width (Previously Defined)  
$ f_{crl} $ = Critical *Local* Buckling Stress

Table C-1-1 may be utilized for determining the Plate Buckling Coefficient,  
$ k $,

### Global Column Stress $ F_n $

Section E2 *Yielding and Global Buckling*

*Compressive Stress*, $ F_n $

When $ \lambda_c \leq 1.5 $:
    $$ F_n = (0.658 ^ {\lambda_{c} ^ 2} ) F_y $$

When $ \lambda_c > 1.5 $:
    $$ F_n = \bigg(\frac{0.877}{\lambda_c^2}\bigg)F_y $$

Where:

$$ \lambda_c = \sqrt{\frac{F_y}{F_{cre}}} $$

$ F_n $ = Global Column Stress  
$ F_{cre} $ = Least of Applicable Global Elastic Buckling Stresses  
(flexural, torsional & flexural-torsional buckling stresses)

### Elastic Global Buckling Stresses $ F_{cre} $

- Determined in accordance with Sections E2.1 through E2.5 of AISI S100-16.
- Section E2.1 applies

#### Section E2.1 - Sections Not Subject to Torsional or Flexural-Torsional Buckling

- Elastic Flexural Buckling Stress, $ F_{cre} $ may be taken as:

$$ F_{cre} = \frac{\pi^2 E}{(\frac{KL}{r})^2} $$

Where:

$K$ = Effective Length Factor determined in accordance with Chapter C  
(Generally = 1.0)  
$L$ = Laterally Unbraced Length of member  
$r$ = Radius of gyration of full unreduced cross-section about axis of  
buckling

Section Capacity = Sum of Individual cross-sectional element capacities

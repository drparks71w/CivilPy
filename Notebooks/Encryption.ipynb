```python
from Crypto.PublicKey import RSA
from Crypto.Cipher import PKCS1_OAEP
from Crypto import Random
# No need for 'ast' anymore

# Generate a 2048-bit RSA key pair
random_generator = Random.new().read
key = RSA.generate(2048, random_generator)
publickey = key.publickey()
```

```python
# 1. Create a cipher object using the public key for encryption
cipher_encrypt = PKCS1_OAEP.new(publickey)
```

```python
# 2. Encrypt the message (must be bytes)
message_bytes = b'encrypt this message'
encrypted = cipher_encrypt.encrypt(message_bytes)

print('encrypted message:', encrypted)
```

```python
# 3. Write the raw bytes to a file in binary mode ('wb')
try:
    with open('encryption.bin', 'wb') as f:
        f.write(encrypted)
    print("Successfully wrote encrypted data to encryption.bin")
except Exception as e:
    print(f"Error writing to file: {e}")
```

```python
# 4. Read the raw bytes from the file in binary mode ('rb')
try:
    with open('encryption.bin', 'rb') as f:
        encrypted_message_from_file = f.read()
    print("Successfully read encrypted data from encryption.bin")

    # 5. Create a cipher object using the *private* key for decryption
    cipher_decrypt = PKCS1_OAEP.new(key)

    # 6. Decrypt the message and decode it back to a string
    decrypted_bytes = cipher_decrypt.decrypt(encrypted_message_from_file)
    decrypted_message = decrypted_bytes.decode('utf-8')

    print('decrypted:', decrypted_message)

except FileNotFoundError:
    print("Error: encryption.bin not found.")
except ValueError:
    print("Error: Decryption failed. Key is incorrect or data is corrupt.")
except Exception as e:
    print(f"An error occurred during decryption: {e}")
```

# Mohr's Circle for a Combined-Stress Element

An invented combined-loading problem: a bracket plate carries
σx = 14 ksi tension, σy = -4 ksi from a clamping force, and 6 ksi shear.
`MohrsCircle` reports the principal stresses and draws both the circle and the
rotated element.

```python
from civilpy.structural.mohrs_circle import MohrsCircle

state = MohrsCircle(sigma_x=14.0, sigma_y=-4.0, tau_xy=6.0)
print(f"center  = {state.center:+.2f} ksi")
print(f"radius  = {state.radius:.2f} ksi")
print(f"sigma_1 = {state.sigma_1:+.2f} ksi")
print(f"sigma_2 = {state.sigma_2:+.2f} ksi")
```

```python
state.plot()
```

```python
# the element rotated to the principal orientation: no shear remains
state.plot_element(theta_deg=state.theta_p_deg)
```

The principal angle lines up the element with the stress trajectories —
the basis for everything from weld checks to the strut directions in a
strut-and-tie model (see the companion STM notebook).

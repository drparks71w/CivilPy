## Why Python is More Prevalent 🧑‍💻

The sheer volume of public-facing content for Python in the GIS world is immense. My training data is a snapshot of the public web, and that's where Python dominates for a few key reasons:

* **Community and Open Source:** The Python GIS community is vast and incredibly active. This results in a massive number of public GitHub repositories, Stack Overflow questions, tutorials, university courses, and blog posts sharing `arcpy` and `ArcGIS API for Python` scripts.
* **Scripting and Data Science:** Python is the go-to language for scripting, automation, and data analysis. GIS analysts, scientists, and researchers frequently share their work as standalone scripts or Jupyter notebooks, creating a huge corpus of examples.
* **Low Barrier to Sharing:** It's much easier and more common for someone to share a 50-line Python script that solves a specific problem than it is to share an entire C# project.

***

## The C# Landscape 🏢

C# is predominantly used for building formal applications and deep integrations, which changes the nature of the available training data:

* **Proprietary Code:** A significant amount of ArcGIS C# development happens within companies to build internal tools or commercial products. This code is typically kept in private repositories and is not part of my public training data.
* **Structured Development:** C# development is more structured, often involving entire Visual Studio projects and solutions. This is less likely to be shared publicly compared to simple scripts.
* **Focused Resources:** While Esri provides excellent, high-quality documentation, sample repositories, and dedicated developer forums for C#, the total volume of informal, community-generated examples is smaller than Python's.

In short, the difference in available training data reflects the real-world usage patterns: **Python for widespread scripting and analysis**, and **C# for more focused application development**.

## Create a New Environment 

by running `conda create -n env_name python=3.10 civilpy` replace `env_name`  
with whatever you want to name your environment, and `3.10` with whichever  
version of python you want to use.

## Install packages

Activate the environment you just created with the command `conda activate civilpy` then install the required packages with `pip install civilpy`

once it's done running the installation script, run `jupyter lab`

if you want to run or edit the notebooks from the civilpy package, first navigate to wherever you downloaded the Notebooks folder with `cd "C:\Path\to\civilpy\Notebooks"` then run `jupyter lab`.

For all future times when you want to run jupyter you no longer have to create or install the packages, just run;

```bash
(base) $ conda activate civilpy
(civilpy) $ jupyter lab
```

to stop the application, just close your browser window and terminal in which it's running.

````python
def get_deck_parameters(span_length: int, continuous_span: bool = False, over_side_drainage: bool = True) -> dict:
    """
    Retrieve bridge deck parameters based on span length and bridge type.
    """
    all_slab_data = continuous_concrete_slab if continuous_span else simple_concrete_slab
    params = all_slab_data[span_length]
    thickness = params['thickness']

    # compatibility flag (not returned)
    U401 = thickness > 20

    top_cover = 2.5
    bot_cover = 1.5
    side_cover = 3

    edge_type = 'over_side_drainage' if over_side_drainage else 'parapet'
    edge_params = params[edge_type]

    result = {
        'span_length': span_length,
        'span_type': 'continuous' if continuous_span else 'simple',
        'thickness': thickness,
        'top_cover': top_cover,
        'bot_cover': bot_cover,
        'side_cover': side_cover,
        'edge_type': edge_type,
    }

    if not continuous_span:
        # Simple: return nested bar dicts directly
        for bar_key in ['a_bar', 'b_bar', 'm_bar', 'n_bar']:
            if bar_key in params:
                result[bar_key] = {
                    'size': params[bar_key]['size'],
                    'spacing': params[bar_key]['spacing'],
                }

        # Edge details (keep D/E metadata as requested)
        result['edge_beam'] = {
            'd': edge_params['d'],
            'x': edge_params['x'],
            'd_bar': {
                'size': edge_params.get('d_bar_size', 0) if over_side_drainage else 0,
                'spacing': edge_params.get('d_bar_num', 0) if over_side_drainage else 0,
            },
            'e_bar': {
                'size': 0 if over_side_drainage else edge_params.get('e_bar_size', 0),
                'spacing': 0 if over_side_drainage else edge_params.get('e_bar_num', 0),
            }
        }
    else:
        # Continuous: mirror the same nested style
        for bar_key in ['a_bar', 'b_bar', 'm_bar', 'n_bar']:
            if bar_key in params:
                # base fields
                bar_obj = {
                    'size': params[bar_key]['size'],
                    'spacing': params[bar_key]['spacing'],
                }
                # optional extras
                if 'a' in params[bar_key]:
                    bar_obj['a'] = params[bar_key]['a']
                if 'length' in params[bar_key]:
                    bar_obj['length'] = params[bar_key]['length']
                if 'no' in params[bar_key]:
                    bar_obj['no'] = params[bar_key]['no']
                result[bar_key] = bar_obj

        # Optional C/D/E bars (include only if present)
        for bar_key in ['c_bar', 'd_bar', 'e_bar']:
            if bar_key in params:
                result[bar_key] = {
                    'size': params[bar_key]['size'],
                    'spacing': params[bar_key]['spacing'],
                    'length': params[bar_key]['length'],
                }

        if 'u_bar_lap' in params:
            result['u_bar_lap'] = params['u_bar_lap']

        # Edge details grouped as in simple, but with F/G/H/J/K nested
        edge_beam = {
            'd': edge_params['d'],
            'x': edge_params['x'],
        }
        for bar_key in ['f_bar', 'g_bar', 'h_bar', 'j_bar', 'k_bar']:
            if bar_key in edge_params:
                bar = edge_params[bar_key]
                eb_entry = {
                    'size': bar['size'],
                    'no': bar.get('no', 0),
                    'length': bar.get('length', 0),
                }
                if bar_key == 'f_bar' and 'f' in bar:
                    eb_entry['f'] = bar['f']
                edge_beam[bar_key] = eb_entry
        result['edge_beam'] = edge_beam

    return result
````

````python
def print_parameters(params: dict) -> None:
    """Pretty print the retrieved parameters."""
    print(f"\n{'='*60}")
    print(f"Bridge Deck Parameters - {params['span_type'].upper()} SPAN")
    print(f"Edge Type: {params['edge_type'].replace('_', ' ').title()}")
    print(f"{'='*60}")
    print(f"Span Length:               {params['span_length']} ft")
    print(f"Deck Thickness:            {params['thickness']} in")
    print(f"Concrete Cover (top/bot/side):  {params['top_cover']} / {params.get('bot_cover', '')} / {params.get('side_cover', '')} in")

    # Determine if bars are nested (simple and new continuous) or flattened (legacy continuous)
    a_bar = params.get('a_bar')
    b_bar = params.get('b_bar')
    m_bar = params.get('m_bar')
    n_bar = params.get('n_bar')

    print(f"\n{'Main Rebar Configuration:':<40}")
    if isinstance(a_bar, dict) and isinstance(b_bar, dict) and isinstance(m_bar, dict) and isinstance(n_bar, dict):
        # Nested structure
        print(f"  Transverse (A) Bar:    #{a_bar['size']} @ {a_bar['spacing']}\" o.c.")
        print(f"  Transverse (B) Bar:    #{b_bar['size']} @ {b_bar['spacing']}\" o.c.")
        print(f"  Longitudinal (M) Bar:  #{m_bar['size']} @ {m_bar['spacing']}\" o.c.")
        print(f"  Longitudinal (N) Bar:  #{n_bar['size']} @ {n_bar['spacing']}\" o.c.")
    else:
        # Backward compatibility with flattened keys
        print(f"  Transverse (A) Bar:    #{params['a_bar_size']} @ {params['a_bar_spacing']}\" o.c.")
        print(f"  Transverse (B) Bar:    #{params['b_bar_size']} @ {params['b_bar_spacing']}\" o.c.")
        print(f"  Longitudinal (M) Bar:  #{params['m_bar_size']} @ {params['m_bar_spacing']}\" o.c.")
        print(f"  Longitudinal (N) Bar:  #{params['n_bar_size']} @ {params['n_bar_spacing']}\" o.c.")
    print(f"{'='*60}\n")
````

````python
def get_aci_bend_radius(bar_size: int) -> float:
    """
    Get the ACI 318 minimum bend diameter *to the centerline of the bar*.
    (ACI specifies *inside* bend radius, so we add 0.5 * bar_diameter)

    Args:
        bar_size: Bar size number (e.g., 8 for #8)

    Returns:
        Centerline bend radius in inches
    """
    bar_diameter = bar_size / 8.0
    inside_bend_diameter = 0.0

    if bar_size >= 3 and bar_size <= 8:
        # ACI 318-14 Table 25.3.2: 6 * db
        inside_bend_diameter = 3 * bar_diameter
    elif bar_size >= 9 and bar_size <= 11:
        # ACI 318-14 Table 25.3.2: 8 * db
        inside_bend_diameter = 8 * bar_diameter
    elif bar_size >= 14 and bar_size <= 18:
        # ACI 318-14 Table 25.3.2: 10 * db
        inside_bend_diameter = 10 * bar_diameter
    else:
        # Default for #2 or other small bars
        inside_bend_diameter = 6 * bar_diameter

    # Return centerline radius
    centerline_bend_diameter = inside_bend_diameter + 0.5 * bar_diameter
    return centerline_bend_diameter
````

````python
def generate_a_bars(span_length,
                    a_bar_size=None,
                    a_bar_spacing=None,
                    deck_thickness=None,
                    deck_width=24.0,
                    bot_cover=1.5,
                    top_cover=2.0,
                    side_cover=3.0,
                    b_bar_size=None): # Added to check clearance
    """
    Generate A-bar geometry (bottom, with hooks).
    PLOTLY (TROUBLESHOOTING) VERSION.
    """
    if a_bar_size is None or a_bar_spacing is None:
        raise ValueError("a_bar_size and a_bar_spacing are required.")
    if deck_thickness is None or top_cover is None or b_bar_size is None:
        raise ValueError("deck_thickness, top_cover, and b_bar_size are required for A-Bar hook clearance.")

    # --- 1. Calculate Hook Geometry ---
    bar_diameter = a_bar_size / 8.0

    # --- START HOOK CLAMPING LOGIC ---
    r_aci = get_aci_bend_radius(a_bar_size)
    b_bar_diameter = b_bar_size / 8.0
    max_z_clearance = deck_thickness - top_cover - b_bar_diameter
    z_bot = bot_cover + bar_diameter / 2.0
    available_hook_height = max_z_clearance - z_bot
    r_max = available_hook_height / 2.0

    if r_aci > r_max:
        print(f"--- WARNING (A-Bar): ACI radius ({r_aci:.3f}\") is too large.")
        print(f"                     Clamping hook radius to max allowed: {r_max:.3f}\"")
        r = r_max
    else:
        print(f"--- INFO (A-Bar): Using ACI hook radius: {r_aci:.3f}\"")
        r = r_aci
    # --- END HOOK CLAMPING LOGIC ---

    ext = max(4 * bar_diameter, 2.5) # Hook extension length
    plot_line_width = bar_diameter * 8.0

    # --- 2. Calculate Bar Positions ---
    span_length_in = span_length * 12.0
    deck_width_in = deck_width * 12.0
    num_bars = int(deck_width_in / a_bar_spacing) + 1

    z_top_of_hook = z_bot + 2 * r # Z-coord of top of 180-deg hook

    # --- 3. Build Plotly Coordinates ---
    x_coords, y_coords, z_coords = [], [], []

    arc_segments = 16 # More segments for smoother arcs

    for i in range(num_bars):
        y = side_cover + (bar_diameter / 2) + i * a_bar_spacing

        # NOTE: Using simplified bar end logic from your version
        L = span_length_in + 10
        bar_end_L_x = 2
        bar_end_R_x = L - 4

        cen_z = z_bot + r # Z-center of the hook arcs

        # --- LEFT HOOK ---
        # L1: Left Extension
        x_coords.extend([bar_end_L_x + r + ext, bar_end_L_x + r])
        y_coords.extend([y, y])
        z_coords.extend([z_top_of_hook, z_top_of_hook])

        # L2: Left Top Arc (90-deg)
        for j in range(arc_segments + 1):
            t = j / arc_segments
            angle = (math.pi / 2) + t * (math.pi / 2) # 90 to 180 deg
            x_coords.append(bar_end_L_x + r * math.cos(angle))
            y_coords.append(y)
            z_coords.append(cen_z + r * math.sin(angle))

        # L3: Left Bottom Arc (90-deg)
        for j in range(arc_segments + 1):
            t = j / arc_segments
            angle = math.pi + t * (math.pi / 2) # 180 to 270 deg
            x_coords.append(bar_end_L_x + r * math.cos(angle))
            y_coords.append(y)
            z_coords.append(cen_z + r * math.sin(angle))

        # --- MAIN STRAIGHT "A" BAR ---
        x_coords.extend([bar_end_L_x, bar_end_R_x])
        y_coords.extend([y, y])
        z_coords.extend([z_bot, z_bot])

        # --- RIGHT HOOK ---
        # R1: Right Bottom Arc (90-deg)
        for j in range(arc_segments + 1):
            t = j / arc_segments
            angle = (3 * math.pi / 2) + t * (math.pi / 2) # 270 to 360 deg
            x_coords.append(bar_end_R_x + r * math.cos(angle))
            y_coords.append(y)
            z_coords.append(cen_z + r * math.sin(angle))

        # R2: Right Top Arc (90-deg)
        for j in range(arc_segments + 1):
            t = j / arc_segments
            angle = 0 + t * (math.pi / 2) # 0 to 90 deg
            x_coords.append(bar_end_R_x + r * math.cos(angle))
            y_coords.append(y)
            z_coords.append(cen_z + r * math.sin(angle))

        # R3: Right Extension
        x_coords.extend([bar_end_R_x - r, bar_end_R_x - ext - r])
        y_coords.extend([y, y])
        z_coords.extend([z_top_of_hook, z_top_of_hook])

        # Break for next bar
        x_coords.append(None); y_coords.append(None); z_coords.append(None)

    # --- 4. Create the Figure ---
    fig = go.Figure()
    fig.add_trace(go.Scatter3d(
        x=x_coords, y=y_coords, z=z_coords, mode='lines',
        name=f'A-Bars (#{a_bar_size})',
        line=dict(color='blue', width=plot_line_width)
    ))
    fig.update_layout(title=f'A-Bar Layout ({num_bars} bars)')
    return fig
````

````python
def generate_b_bars(span_length,
                    b_bar_size=None,
                    b_bar_spacing=None,
                    deck_thickness=None, # NOW REQUIRED
                    deck_width=24.0,
                    top_cover=2.0,
                    side_cover=3.0):
    """
    Generate B-bar geometry (top, straight, transverse X-dir).
    PLOTLY (TROUBLESHOOTING) VERSION.
    """
    if b_bar_size is None or b_bar_spacing is None:
        raise ValueError("b_bar_size and b_bar_spacing are required.")
    if deck_thickness is None:
        raise ValueError("deck_thickness is required to calculate B-bar elevation.")

    # --- 1. Calculate Bar Geometry ---
    bar_diameter = b_bar_size / 8.0
    plot_line_width = bar_diameter * 8.0

    # --- 2. Calculate Bar Positions ---
    span_length_in = span_length * 12.0
    deck_width_in = deck_width * 12.0
    num_bars = int(deck_width_in / b_bar_spacing) + 1

    # Z-coord of the *center* of the top B-bar
    z_pos = deck_thickness - top_cover - (bar_diameter / 2.0)

    # --- 3. Build Plotly Coordinates ---
    x_coords, y_coords, z_coords = [], [], []

    for i in range(num_bars):
        y = side_cover + (bar_diameter / 2) + i * b_bar_spacing
        L = span_length_in + 10
        bar_end_L_x = 2
        bar_end_R_x = L - 4

        # --- MAIN STRAIGHT "B" BAR ---
        x_coords.extend([bar_end_L_x, bar_end_R_x])
        y_coords.extend([y, y])
        z_coords.extend([z_pos, z_pos]) # Use the new z_pos

        x_coords.append(None); y_coords.append(None); z_coords.append(None)

    # --- 4. Create the Figure ---
    fig = go.Figure()
    fig.add_trace(go.Scatter3d(
        x=x_coords, y=y_coords, z=z_coords, mode='lines',
        name=f'B-Bars (#{b_bar_size})',
        line=dict(color='red', width=plot_line_width)
    ))
    fig.update_layout(title=f'B-Bar Layout ({num_bars} bars)')
    return fig
````

````python
def generate_m_bars(span_length,
                    m_bar_size=None,
                    m_bar_spacing=None,
                    deck_width=24.0,
                    bot_cover=1.5,
                    side_cover=3.0,
                    a_bar_size=None):
    """
    Generate M-bar geometry (bottom, straight, longitudinal Y-dir).
    PLOTLY (TROUBLESHOOTING) VERSION.
    """
    if m_bar_size is None or m_bar_spacing is None:
        raise ValueError("m_bar_size and m_bar_spacing are required.")
    if a_bar_size is None:
        raise ValueError("a_bar_size is required to calculate M-bar elevation.")

    # --- 1. Calculate Bar Geometry ---
    bar_diameter = m_bar_size / 8.0
    a_bar_diameter = a_bar_size / 8.0
    plot_line_width = bar_diameter * 8.0

    # --- 2. Calculate Bar Positions ---
    span_length_in = span_length * 12.0
    deck_width_in = deck_width * 12.0

    # Z-coord: Center of M-bar is just above top of A-bar
    z_pos = bot_cover + a_bar_diameter + (bar_diameter / 2.0)

    # Use X-direction end/side cover logic from A/B bars
    L_span = span_length_in + 10
    bar_end_L_x = 2
    bar_end_R_x = L_span - 4

    # Use Y-direction side cover
    y_start = side_cover
    y_end = deck_width_in - side_cover

    # Bars are spaced along the X-axis
    num_bars = int((bar_end_R_x - bar_end_L_x) / m_bar_spacing) + 1

    # --- 3. Build Plotly Coordinates ---
    x_coords, y_coords, z_coords = [], [], []

    for i in range(num_bars):
        # Each bar is at a constant X, spaced by m_bar_spacing
        x = bar_end_L_x + i * m_bar_spacing

        # --- MAIN STRAIGHT "M" BAR (Y-direction) ---
        x_coords.extend([x, x]) # X is constant
        y_coords.extend([y_start, y_end]) # Y changes
        z_coords.extend([z_pos, z_pos]) # Z is constant

        x_coords.append(None); y_coords.append(None); z_coords.append(None)

    # --- 4. Create the Figure ---
    fig = go.Figure()
    fig.add_trace(go.Scatter3d(
        x=x_coords, y=y_coords, z=z_coords, mode='lines',
        name=f'M-Bars (#{m_bar_size})',
        line=dict(color='green', width=plot_line_width)
    ))
    fig.update_layout(title=f'M-Bar Layout ({num_bars} bars)')
    return fig


def generate_n_bars(span_length,
                    n_bar_size=None,
                    n_bar_spacing=None,
                    deck_thickness=None,
                    deck_width=24.0,
                    top_cover=2.0,
                    side_cover=3.0,
                    b_bar_size=None):
    """
    Generate N-bar geometry (top, straight, longitudinal Y-dir).
    PLOTLY (TROUBLESHOOTING) VERSION.
    """
    if n_bar_size is None or n_bar_spacing is None:
        raise ValueError("n_bar_size and n_bar_spacing are required.")
    if deck_thickness is None or b_bar_size is None:
        raise ValueError("deck_thickness and b_bar_size are required to calculate N-bar elevation.")

    # --- 1. Calculate Bar Geometry ---
    bar_diameter = n_bar_size / 8.0
    b_bar_diameter = b_bar_size / 8.0
    plot_line_width = bar_diameter * 8.0

    # --- 2. Calculate Bar Positions ---
    span_length_in = span_length * 12.0
    deck_width_in = deck_width * 12.0

    # Z-coord: Center of N-bar is just below bottom of B-bar
    z_b_bar_bottom = (deck_thickness - top_cover) - b_bar_diameter
    z_pos = z_b_bar_bottom - (bar_diameter / 2.0)

    # Use X-direction end/side cover logic from A/B bars
    L_span = span_length_in + 10
    bar_end_L_x = 2
    bar_end_R_x = L_span - 4

    # Use Y-direction side cover
    y_start = side_cover
    y_end = deck_width_in - side_cover

    # Bars are spaced along the X-axis
    num_bars = int((bar_end_R_x - bar_end_L_x) / n_bar_spacing) + 1

    # --- 3. Build Plotly Coordinates ---
    x_coords, y_coords, z_coords = [], [], []

    for i in range(num_bars):
        # Each bar is at a constant X, spaced by n_bar_spacing
        x = bar_end_L_x + i * n_bar_spacing

        # --- MAIN STRAIGHT "N" BAR (Y-direction) ---
        x_coords.extend([x, x]) # X is constant
        y_coords.extend([y_start, y_end]) # Y changes
        z_coords.extend([z_pos, z_pos]) # Z is constant

        x_coords.append(None); y_coords.append(None); z_coords.append(None)

    # --- 4. Create the Figure ---
    fig = go.Figure()
    fig.add_trace(go.Scatter3d(
        x=x_coords, y=y_coords, z=z_coords, mode='lines',
        name=f'N-Bars (#{n_bar_size})',
        line=dict(color='purple', width=plot_line_width)
    ))
    fig.update_layout(title=f'N-Bar Layout ({num_bars} bars)')
    return fig
````
